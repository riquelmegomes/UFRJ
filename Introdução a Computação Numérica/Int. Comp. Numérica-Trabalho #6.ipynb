## Tarefa #1:

Desenvolver códigos, na forma de funções, que implementem os seguintes métodos:

a) Método da bissecção. 

b) Método de Newton-Raphson.  

c) Método da secante.

Sugestão: implemente os métodos na forma de funções que devolvem na saída uma mensagem de erro ou o valor aproximado do zero procurado. Evite a possibilidade de laços infinitos considerando um número máximo de iterações como parâmetro de entrada.

````python
# a) Função que implementa o método da bissecção

def bisseccao(f, a, b, tol, Niter):

  k = 1
  fa = f(a)

  while(k <= Niter):
    p = a + ((b - a) / 2)
    fp = f(p)
    if(fp == 0 or ((b - a) / 2) < tol):
      return p, k
    if(fa * fp > 0):
      a = p
      fa = fp
    else:
      b = p

    k = k + 1
  
  print("Tolerância não atingida")
````

````python
# b) Função que implementa o método de Newton-Raphson

def newton(f, derivada, p0, tol, Niter):
  
  k = 1
  while(k <= Niter):
    p = p0 - (f(p0) / derivada(p0))

    if(abs(p - p0) < tol):
      return p, k
    
    p0 = p
    k = k + 1
  
  print("Tolerância não atingida")
````

````python
# c) Função que implementa o método da secante

def secante(f, p0, p1, tol, Niter):
  
  k = 1
  while(k <= Niter):
    p = p1 - ((f(p1)*(p1 - p0)) / (f(p1) - f(p0)))

    if(abs(p - p1) < tol):
      return p, k
    
    p0 = p1
    p1 = p
    k = k + 1
  
  print("Tolerância não atingida")
````

## Tarefa #2: 

a) A equação de van der Waals 
           		$$\left(p+\frac{a}{v^2} \right)(v-b)=R T,$$

onde $p$ é a pressão, $v$ representa o volume molar, $a$ e $b$ são constantes empíricas que dependem do gás,  $R$ representa a constante universal dos gases e $T$ é a temperatura absoluta. 
Determine os volumes molares do dióxido de carbono a uma temperatura  absoluta $T=300\,K$ e pressão de $1$, $10$, $50$ e $100 \,atm$, respectivamente. 
Os valores das constantes são $R=0.082054 \,L\, atm/(mol K)$, $a=3.592 \,L^2 atm/mol^2$ e $b=0.04267 L/mol$.

Use dois dos métodos implementados considerando uma tolerância $\varepsilon=5\cdot10^{-6}$.

b) Compare os resultados dos dois métodos usados, considerando os valores obtidos e a quantidade de iterações realizadas. Escreva suas observações no relatório.


````python
# a)

pressoes = [1, 10, 50, 100]  # valor da pressão
a, b = 3.592, 0.04267  # constantes empíricas
R = 0.082054  # constante universal dos gases
T = 300  # temperatura absoluta

def van_der_waals(v):
  return (p + (a / v**2)) * (v - b) - (R * T)

tol = 5 * (10**(-6))
intervalo = [0.01, 30]
Niter = 25

for i in range(len(pressoes)):
  p = pressoes[i]
  vb, kb = bisseccao(van_der_waals, intervalo[0], intervalo[1], tol, Niter)
  p0 = vb
  p1 = vb + 1
  vs, ks = secante(van_der_waals, p0, p1, tol, Niter)

  print(f"PRESSÃO DE {p} ATM:")
  print(f"[Bissecção] Volume molar: {vb}\n[Secante] Volume molar: {vs}\n")
````

````python
# b)

for i in range(len(pressoes)):
  p = pressoes[i]
  vb, kb = bisseccao(van_der_waals, intervalo[0], intervalo[1], tol, Niter)
  p0 = vb
  p1 = vb + 1
  vs, ks = secante(van_der_waals, p0, p1, tol, Niter)

  print(f"PRESSÃO DE {p} ATM:")
  print(f"[Bissecção] Volume molar: {vb}\tNumero de iterações: {kb}")
  print(f"[Secante] Volume molar: {vs}\tNumero de iterações: {ks}")
  print(f"Módulo da diferença: {abs(vb - vs)}\n")
````

Observe que o número $\pi$ é um zero das funções:

i) $f(x)= \cos(x)+1$

ii) $f(x)= \mathrm{sen}(x)$

a) Use sua implementação do método de Newton-Raphson para obter aproximações de considerando as funções acima e uma tolerância $\varepsilon =10^{-16}$.

b) Compare os resultados dos itens i) e ii), considerando os erros das aproximações obtidas com relação ao valor "exato" dado pelo Python no módulo math e a quantidade de iterações efetuadas. Escreva suas observações no relatório.

c) Explique o que foi observado no item b) no seu relatório.

````python
# a)
import math 

# i)
def f(x):
  return math.cos(x) + 1

def derivada_f(x):
  return -1*math.sin(x)

# ii)
def f2(x):
  return math.sin(x)

def derivada_f2(x):
  return math.cos(x) 

p0 = 3.14
tol = math.pow(10, -16)
Niter = 20
aproximacoes = []
qtd_iteracoes = []

p, k = newton(f, derivada_f, p0, tol, Niter)
aproximacoes.append(p)
qtd_iteracoes.append(k)
print(f"Função i)\nAproximação de pi: {p}\n")

p, k = newton(f2, derivada_f2, p0, tol, Niter)
aproximacoes.append(p)
qtd_iteracoes.append(k)
print(f"Função ii)\nAproximação de pi: {p}")
````

````python
# b) 
import math

valor_exato = math.pi
for i in range(len(aproximacoes)):
  valor_aproximado = aproximacoes[i]
  
  print(f"Aproximação de pi: {valor_aproximado}\nValor exato de pi: {valor_exato}")
  print(f"Erro absoluto: {abs(valor_exato - valor_aproximado)}")
  print(f"Numero de iterações: {qtd_iteracoes[i]}\n")
````
