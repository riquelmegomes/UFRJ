## Tarefa #1

Implemente funções em Python para aproximar a integral definida
$$
I = \int_a^b f(x)\,dx,
$$
usando: 
i) a regra dos trapézios repetida/composta;
ii) a regra de Simpson repetida/composta.

a) Considere como parâmetros de entrada o espaçamento $h$ entre os nós (pontos) que serão usados e uma lista com os valores da função nesses nós: $f_j = f(a+j\, h)$,  $j=0,\dots,n$.

b) Considere como parâmetros de entrada o intervalo $[a, b]$, a função $f$ e a quantidade de nós (pontos) que serão usados.

**Observação**: Você precisará implementar 4 funções.


````python
# a)

def trapezio_rep1(h, lista_valores):
  n = (len(lista_valores) - 1)
  f_a = lista_valores[0]
  f_b = lista_valores[n]
  somatorio = 0

  for j in range(1, n):
    somatorio += lista_valores[j]
  
  integral = (h / 2) * (f_a + f_b + 2*somatorio)
  return integral


def simpson_rep1(h, lista_valores):
  n = (len(lista_valores) - 1)
  f_a = lista_valores[0]
  f_b = lista_valores[n]
  somatorio1, somatorio2 = 0, 0

  for j in range(1, (n/2)+1):
    if(j != n/2): somatorio1 += lista_valores[2*j]
    somatorio2 += lista_valores[2*j - 1]

  integral = (h / 3) * (f_a + f_b + 2*somatorio1 + 4*somatorio2)
  return integral
````

````python
# b)

def trapezio_rep2(intervalo, f, qtd_nos):
  n = (qtd_nos - 1)
  a = intervalo[0]
  b = intervalo[1]
  h = (b - a) / n
  somatorio = 0

  for j in range(1, n):
    x = a + j*h
    somatorio += f(x)

  integral = (h / 2) * (f(a) + f(b) + 2*somatorio)
  return integral


def simpson_rep2(intervalo, f, qtd_nos):
  n = (qtd_nos - 1)
  a = intervalo[0]
  b = intervalo[1]
  h = (b - a) / n
  somatorio1, somatorio2 = 0, 0

  for j in range(1, int(n/2)+1):
    if(j != n/2): 
      x = a + (2*j)*h
      somatorio1 += f(x)
    x = a + (2*j - 1)*h
    somatorio2 += f(x)

  integral = (h / 3) * (f(a) + f(b) + 2*somatorio1 + 4*somatorio2)
  return integral
````

A **função erro** é definida como
$$
\mathrm{erf}(x) = \frac{2}{\sqrt{\pi}}\int_0^x \mathrm{e}^{-t^2}\,dt.
$$

a) Implemente uma função que determine o valor aproximado de $\mathrm{erf}(x)$ para qualquer entrada $x$ com um erro absoluto menor que $10^{-8}$. Explique as ideias básicas da sua implementação.

b) Teste a função implementada comparando com os seguintes valores*: 

i) $\mathrm{erf}(-2) = -0.9953222650189527$, 

ii) $\mathrm{erf}(0.5) = 0.5204998778130465$, 

iii) $\mathrm{erf}(1) = 0.8427007929497149$, 

iv) $\mathrm{erf}(5) = 0.9999999999984625$.

*Os valores apresentados foram calculados com 16 dígitos de precisão.

````python
# a) 
import math

def funcao_f(t):
  return math.pow(math.e, -1*math.pow(t, 2))

# função erro
def aprox_erf(x):
  intervalo = [0, x]
  qtd_nos = 1000001
  resultado = (2 / math.sqrt(math.pi)) * simpson_rep2(intervalo, funcao_f, qtd_nos)
  
  return resultado
````

````python
# b)
valores = [-0.9953222650189527, 0.5204998778130465, 0.8427007929497149, 0.9999999999984625]
entradas = [-2, 0.5, 1, 5]

for k in range(len(entradas)):
  valor_exato = valores[k]
  x = entradas[k]
  resultado = aprox_erf(x)
  print(f"x = {x}")
  print(f"Valor exato: {valor_exato} \t aprox_erf({x}): {resultado} \t\t Erro absoluto: {abs(resultado - valor_exato)}\n")
````
