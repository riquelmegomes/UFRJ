# **Sistemas dinâmicos**

> #### Sistemas dinâmicos são um conceito importante em várias disciplinas, incluindo matemática, física, biologia, economia e engenharia. Esses sistemas são chamados de "dinâmicos" porque envolvem elementos que mudam com o tempo, e o comportamento do sistema é frequentemente caracterizado por suas trajetórias ou evolução ao longo do tempo. O conceito de sistema dinâmico nasce da exigência de construir um modelo geral para os sistemas físicos que evoluem no tempo, segundo uma regra que liga o estado presente aos estados passados.

<br>

#### Aqui estão algumas características essenciais dos sistemas dinâmicos:
<br>

1. #### **Estado**: Um sistema dinâmico é descrito por seu estado, que é uma coleção de variáveis que representam as condições do sistema em um determinado momento. O estado pode ser representado por um vetor de variáveis ​​que evoluem ao longo do tempo.

2. #### **Evolução temporal**: Os sistemas dinâmicos evoluem ao longo do tempo de acordo com regras ou equações específicas que descrevem como o estado do sistema muda com o tempo. Essas equações diferenciais são frequentemente usadas para modelar sistemas dinâmicos contínuos, enquanto sistemas discretos podem ser modelados por equações de diferenças.

3. #### **Trajetórias**: As trajetórias são as sequências de estados que um sistema dinâmico percorre ao longo do tempo. O comportamento do sistema é muitas vezes representado visualmente por meio de gráficos que mostram como o estado evolui com o tempo.

4. #### **Atratores**: Em muitos sistemas dinâmicos, existem pontos ou regiões no espaço de estado para os quais o sistema tende a convergir à medida que o tempo passa. Esses pontos ou regiões são chamados de atratores e podem ser estáveis (o sistema converge para eles) ou instáveis (o sistema se afasta deles).

5. #### **Caos**: Alguns sistemas dinâmicos exibem comportamento caótico, o que significa que pequenas variações nas condições iniciais podem levar a resultados drasticamente diferentes ao longo do tempo. O estudo do caos é um campo fascinante na teoria dos sistemas dinâmicos.

6. #### **Estabilidade e equilíbrio**: Em muitos sistemas dinâmicos, os pesquisadores estão interessados em identificar pontos de equilíbrio, onde as taxas de mudança de todas as variáveis ​​são zero. A estabilidade desses pontos de equilíbrio é um tópico importante, pois determina como o sistema responde a perturbações.

7. #### **Aplicações**: Sistemas dinâmicos são usados para modelar uma ampla variedade de fenômenos do mundo real, desde o comportamento de populações de animais até a dinâmica de mercados financeiros. Eles também são usados na engenharia para controlar sistemas complexos, como aeronaves e sistemas de automação.

<br>

#### Resumidamente, sistemas dinâmicos são uma maneira de modelar e entender como sistemas complexos evoluem ao longo do tempo. Eles desempenham um papel fundamental na compreensão de muitos fenômenos naturais e artificiais e têm uma ampla gama de aplicações em diversas disciplinas.

## **Sistemas dinâmicos autônomos**

> #### Um sistema dinâmico autônomo é um tipo específico de sistema dinâmico em que as equações que descrevem sua evolução não dependem explicitamente do tempo. Em outras palavras, é um sistema cujo comportamento não é afetado por uma variável de tempo externa.

<br>

#### Exemplos de sistemas dinâmicos autônomos incluem o pêndulo simples (quando não há forças externas ou forçamento periódico), sistemas de equações diferenciais autônomas que descrevem a dinâmica de populações biológicas em equilíbrio, ou mesmo sistemas elétricos passivos, onde os componentes não mudam com o tempo.

<br>

#### Os sistemas dinâmicos autônomos são fundamentais em muitos campos da ciência e da engenharia, pois sua simplicidade matemática e a capacidade de prever seu comportamento com base em condições iniciais os tornam um objeto de estudo valioso para a compreensão de sistemas complexos.

## **Exemplos**

#### Sistemas dinâmicos estão presentes em muitos aspectos da vida cotidiana e na ciência. Aqui estão alguns exemplos de sistemas dinâmicos:

1. #### **Pêndulo simples**: Um pêndulo oscilante é um exemplo clássico de um sistema dinâmico. A posição angular do pêndulo muda continuamente com o tempo de acordo com as leis da física.

2. #### **Populações biológicas**: Modelar o crescimento de populações de animais ou plantas é um exemplo de um sistema dinâmico. As interações entre reprodução, recursos e predação podem levar a padrões complexos de crescimento populacional.

3. #### **Tráfego rodoviário**: O tráfego em uma rede de estradas é um sistema dinâmico complexo. Os padrões de fluxo de tráfego podem mudar ao longo do tempo devido ao comportamento dos motoristas e a eventos como congestionamentos.

4. #### **Dinâmica de fluidos**: A dinâmica de fluidos estuda o movimento de líquidos e gases. É usado para modelar fenômenos como o movimento da água em rios, o comportamento do ar em torno de aeronaves e a previsão do tempo.

5. #### **Mercados financeiros**: Os mercados de ações, moedas e commodities são sistemas dinâmicos onde os preços e as transações mudam continuamente. Os investidores tentam prever as mudanças de preço usando análise técnica e fundamental.



### **Sistema de Lorenz**

> #### O Sistema de Lorenz, nomeado após o meteorologista Edward Lorenz, é um conjunto de equações diferenciais não lineares que descrevem o comportamento caótico de um **sistema dinâmico**. Esse sistema é amplamente reconhecido por ser um dos primeiros exemplos de um sistema determinístico que exibe sensibilidade extrema às condições iniciais, o que leva ao comportamento caótico.

\
#### As equações de Lorenz foram originalmente formuladas para modelar a convecção atmosférica simplificada mas, ao longo do tempo, elas foram adotadas para descrever uma variedade de fenômenos, desde sistemas climáticos até oscilações químicas e físicas. O sistema consiste em três equações diferenciais acopladas, que representam as variações temporais de três variáveis dependentes: $x$, $y$ e $z$.


#### As equações de Lorenz são as seguintes:
1. ### $dx/dt = σ(y - x)$
2. ### $dy/dt = x(ρ - z) - y$
3. ### $dz/dt = xy - βz$

#### Onde $x$, $y$ e $z$ são as variáveis do sistema. $σ$, $ρ$ e $β$ são os parâmetros do sistema, que determinam o comportamento do sistema e a sua sensibilidade às condições iniciais e t é o tempo.
<br>

#### Uma das características mais marcantes do sistema de Lorenz é o fenômeno conhecido como "efeito borboleta". Isso se refere à sensibilidade extrema às condições iniciais, em que uma pequena alteração nos valores iniciais das variáveis pode resultar em comportamentos completamente diferentes no longo prazo. Devido à natureza caótica do sistema de Lorenz, ele não é previsível no longo prazo, mesmo que as equações que o descrevem sejam determinísticas e completamente conhecidas. Isso significa que, para sistemas caóticos como esse, pequenas incertezas nas medições iniciais ou nos parâmetros podem resultar em predições substancialmente diferentes ao longo do tempo.
<br>

#### Em resumo, o sistema de Lorenz é um exemplo clássico de sistema dinâmico caótico que ajudou a estabelecer as bases da teoria do caos. Suas equações ilustram como sistemas aparentemente simples podem exibir comportamento complexo e imprevisível devido à sua sensibilidade extrema às condições iniciais.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Aug 13 21:51:58 2023

@author: amaury
"""

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definição das equações diferenciais do sistema de Lorenz
def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Condições iniciais
initial_conditions = [0, 1, 1.05]

# Intervalo de tempo para a integração
t_span = (0, 25)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Parâmetros do sistema de Lorenz
sigma = 10
rho = 28
beta = 8/3

# Resolvendo o sistema de Lorenz
solution = solve_ivp(lorenz, t_span, initial_conditions, args=(sigma, rho, beta), t_eval=t_eval)

# Plotando o gráfico da solução
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2], lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f'Solução do Sistema de Lorenz\nCondições iniciais: {initial_conditions}')
plt.show()
```

Acima, o gráfico construído como resultado do código do sistema de Lorenz com condições iniciais $[0, 1, 1.05]$ e parâmetros $σ = 10$, $ρ = 28$ e $β = 8/3$, representando visualmente uma curva que aparenta circundar ao redor de dois pontos centrais, os atratores de Lorenz.

# **Aproximação de um sistema de dados**

> #### A aproximação de um sistema de dados é um processo fundamental em várias áreas, incluindo matemática, estatística, ciência da computação e engenharia. Refere-se ao ato de criar um modelo ou representação simplificada de um sistema complexo de dados com o objetivo de compreendê-lo melhor, fazer previsões ou tomar decisões informadas. Essa abordagem é amplamente utilizada em situações em que o sistema real é intrincado demais para ser analisado diretamente ou quando se deseja simplificar a complexidade para tornar a análise mais viável.

<br>

#### Com isso, a aproximação de um sistema de dados é uma ferramenta poderosa para lidar com sistemas complexos, permitindo a compreensão, previsão e tomada de decisões informadas em uma variedade de campos e setores. É um processo iterativo que envolve a escolha e o ajuste de modelos, e sua eficácia depende da qualidade dos dados disponíveis e da capacidade de validação do modelo escolhido.

## **Algoritmo de aproximação de um sistema de dados baseado no modelo de Lorenz**

#### Para descrever o algoritmo de aproximação de um sistema de dados baseado no modelo de Lorenz, o qual vai simular dados reais, podemos seguir os seguintes passos:

#### Definir a função do sistema de equações diferenciais de Lorenz, que gera os dados do modelo. No nosso caso, podemos usar os valores de $β$, $σ$ e $ρ$ fornecidos e definir a função: $lorenz(xyz, t0, sigma = σ, beta = β, rho = ρ)$, que recebe um vetor $xyz$ com as coordenadas $x$, $y$ e $z$ do sistema e o tempo $t0$ como entrada e retorna as derivadas $dx/dt$, $dy/dt$ e $dz/dt$.



```python
beta = 8/3
sigma = 10
rho =30

# Define the 3D ODE system
def lorenz(x_y_z, t0, sigma=sigma, beta=beta, rho=rho):
    x, y, z = x_y_z
    return [sigma*(y-x), x*(rho-z)-y, x*y-beta*z]
```

#### Gerar os dados do modelo de Lorenz usando a função $lorenz()$ e armazenar os resultados em uma matriz de dados. No exemplo dado, a matriz de dados é chamada de $”data”$ e tem três colunas correspondentes às coordenadas $x$, $y$ e $z$ do sistema.

```python
import numpy as np
from scipy import integrate

x0 = (0, 1, 20)
dt = 0.001
t = np.arange(0, 50+dt, dt)
x_t = integrate.odeint(lorenz, x0, t,rtol=10**(-12),atol=10**(-12)*np.ones_like(x0))
x, y, z = x_t.T

data_r = np.column_stack([x, y, z])
data = data_r.T
print(data.shape)
```

#### Aplicar o algoritmo de redução de dimensão usando a técnica de Análise de Componentes Principais (PCA) ou SVD (Decomposiçãoo de Valores Singulares).

<br>

#### No exemplo dado, o algoritmo de redução de dimensão é a função $CamR(X, X′, r)$, que recebe a matriz de dados $data$ e constrói duas matrizes $X$ e $X′$, onde $X$ é a matriz de dados original e $X'$ é a matriz de dados deslocada uma posição para a direita. O parâmetro $r$ é a dimensão reduzida que desejamos. A função $CamR()$ executa os seguintes passos:
* #### Aplica a decomposição SVD na matriz $X$.
* #### Calcula as matrizes reduzidas $Ur$, $Sigma\_r$ e $VT\_r$ com base na dimensão $r$ desejada.
* #### Calcula a matriz $Â$, que é a aproximação linear da matriz $X′$ usando a matriz pseudo-inversa de $X$.
* #### Retorna a matriz de valores singulares $Sigma$ e a matriz $Â$.


```python
def CamR(X, Xprime, r):

    # SVD
    U, Sigma, VT = np.linalg.svd(X, full_matrices=0)

    # Redução de dimensão
    U_r = U[:, :r]
    Sigma_r = np.diag(Sigma[:r])
    VT_r = VT[:r, :]

    # Cálculo da Â = Xprime*pseudo-inversa(X)
    Â = np.linalg.solve(Sigma_r.T, (U_r.T @ Xprime @ VT_r.T).T).T

    # Cálculo da matriz reconstruída
    Xr = U_r @ Sigma_r @ VT_r

    return Sigma, Â, Xr
```

#### Usar a matriz Â para reconstruir os dados originais em uma dimensão reduzida. Isso é feito calculando a matriz reconstruída $Xr$ como sendo igual a multiplicacão das matrizes $Ur$, $Sigma\_r$ e $VT\_r$ transpostas. No exemplo dado, podemos obter a matriz reconstruída com a seguinte equação: $Xr = Ur @Sigma\_r@VT\_r.T$

```python
# Crie matrix1 da primeira à penúltima coluna de dados
X = data[:, :-1]

# Crie matrix2 da segunda à última coluna de dado
Xprime = data[:, 1:]

print(X.shape)
print(Xprime.shape)

Sigma, Â, Xr = CamR(X, Xprime, r=21)

# Matriz reconstruída
print(Xr.shape)
```

#### Comparar a matriz reconstruída $Xr$ com a matriz original $X$ e avaliar a precisão da aproximação. Isso pode ser feito usando métricas de erro, como o erro quadrático médio (RMSE) ou a diferença absoluta média (MAD).

**Erro quadrático médio (RMSE)**

O Erro Quadrático Médio (RMSE - Root Mean Square Error) é uma métrica amplamente usada para avaliar a qualidade de um modelo de previsão ou uma aproximação em relação aos valores reais ou alvos. O RMSE mede a média das diferenças ao quadrado entre os valores previstos pelo modelo e os valores reais e, em seguida, extrai a raiz quadrada desse valor médio. Ele fornece uma medida da dispersão ou erro típico entre as previsões e os valores reais.

<br>

A fórmula do RMSE é a seguinte: $RMSE = √(Σ(y_i - ŷ_ i)^2 / n)$ onde:
* $y_i$ representa o valor real da i-ésima observação.
* $ŷ_i$ representa o valor previsto pelo modelo para a i-ésima observação.
* $Σ$ representa a soma sobre todas as observações.
* $n$ é o número total de observações.


```python
import numpy as np

def calcular_rmse(matriz_real, matriz_aproximada):

    # Verificar se as matrizes têm as mesmas dimensões
    if matriz_real.shape != matriz_aproximada.shape:
        raise ValueError("As matrizes devem ter as mesmas dimensões.")

    # Calcular o erro quadrático médio
    erro_quadratico = np.square(matriz_real - matriz_aproximada)
    rmse = np.sqrt(np.mean(erro_quadratico))

    print("Erro Quadrático Médio (RMSE):", rmse)

calcular_rmse(X, Xr)
```

**Diferença absoluta média (MAD)**

A Diferença Absoluta Média (MAD - Mean Absolute Difference) é uma métrica usada para avaliar a precisão ou a qualidade de uma previsão ou estimativa em relação aos valores reais. A MAD mede a média das diferenças absolutas entre os valores previstos (ou estimados) e os valores reais. É uma métrica que fornece uma medida do erro médio, levando em consideração apenas a magnitude das discrepâncias, independentemente da direção (positiva ou negativa) das diferenças.

<br>

A fórmula da MAD é a seguinte: $MAD = (1/n) * Σ|yi - ŷi|$ onde:
* $y_i$ representa o valor real da i-ésima observação.
* $ŷ_i$ representa o valor previsto pelo modelo para a i-ésima observação.
* $Σ$ representa a soma sobre todas as observações.
* $n$ é o número total de observações.

```python
import numpy as np

def calcular_mad(matriz_real, matriz_aproximada):

    # Verificar se as matrizes têm as mesmas dimensões
    if matriz_real.shape != matriz_aproximada.shape:
        raise ValueError("As matrizes devem ter as mesmas dimensões.")

    # Calcular a diferença absoluta entre as matrizes
    diferenca_absoluta = np.abs(matriz_real - matriz_aproximada)

    # Calcular a MAD (Média da Diferença Absoluta)
    mad = np.mean(diferenca_absoluta)

    print("Diferença Absoluta Média (MAD):", mad)

calcular_mad(X, Xr)
```

Podemos perceber que, com ambas métricas de erro, o valor do erro encontrado entre a matriz original $X$ e a matriz reconstruída $Xr$ é muito pequeno. Isso indica que a nossa aproximação apresenta uma precisão consideravelmente boa.
