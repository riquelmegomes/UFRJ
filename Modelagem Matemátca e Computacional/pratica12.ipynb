# **Processos Estocásticos**

> #### Processos estocásticos referem-se a uma classe de fenômenos ou sistemas que evoluem ao longo do tempo de maneira probabilística, ou seja, envolvem algum grau de aleatoriedade. Esses processos são amplamente utilizados para modelar e entender sistemas dinâmicos nos quais a evolução ao longo do tempo não pode ser completamente determinada, seja devido a variabilidade inerente, ruído ou outras fontes de incerteza.

<br>

#### Existem dois tipos principais de processos estocásticos: discretos e contínuos.

<br>

### **Processos Estocásticos Discretos**

* #### **Cadeias de Markov:** São processos nos quais um sistema assume diferentes estados em intervalos discretos de tempo, e a transição entre estados é governada por probabilidades.

* #### **Processos de Bernoulli:** Envolve eventos binários, como sucesso ou fracasso, onde a probabilidade de sucesso ou fracasso pode variar.

<br>

#### **Processos Estocásticos Contínuos:**

* #### **Processos de Wiener (ou Processos Gaussianos):** Modelam fenômenos contínuos e são frequentemente utilizados em finanças e física para descrever movimentos brownianos.

* #### **Processos de Poisson:** Modelam a ocorrência de eventos em um intervalo contínuo de tempo ou espaço, sendo úteis em áreas como telecomunicações, biologia e filas de espera.

<br>

#### A teoria dos processos estocásticos é fundamental em estatística, matemática aplicada, economia, engenharia, entre outras disciplinas. Eles oferecem uma maneira poderosa de lidar com a incerteza e são usados em uma variedade de contextos, desde previsões financeiras até modelagem de sistemas biológicos.

# **Equação de Langevin**

#### A equação de Langevin é uma equação estocástica que descreve o movimento de partículas submetidas a forças aleatórias. A equação de Langevin pode ser descrita como: $$ m\frac{dv}{dt} = -γv + F + η(t)$$ onde:

* #### $m$ é a massa da partícula
* #### $v$ é a velocidade da partícula
* #### $t$ é o tempo
* #### $γ$ é coeficiente de atrito
* #### $F$ é a força externa
* #### $η(t)$ é a força aleatória ou ruído térmico

<br>

## **Interpretação dos termos**

* #### $m\frac{dv}{dt}$ é a taxa de variação da quantidade de movimento.
* #### $-γv$ é a força de arrasto
* #### $F$ é a força externa aplicada
* #### $η(t)$ é a força aleatória devido ao movimento térmico

<br>

## **Solução da Equação de Langevin**


#### A solução da equação de Langevin depende das propriedades do ruído térmico η(t) e das forças externas aplicadas.
* #### Para forças externas simples e ruído térmico branco, a equação pode ser resolvida analiticamente.
* #### Para casos mais complexos, é necessário recorrer a métodos numéricos.

<br>

## **Aplicações da Equação de Langevin**

#### A equação de Langevin é utilizada em diversas áreas da física, tais como:
* #### Movimento browniano de partículas suspensas em um fluido.
* #### Dinâmica de partículas em campos magnéticos.
* #### Comportamento de polímeros e macromoléculas.


## **Solução da Equação de Langevin com RK2**



### **Método Runge-Kutta de Ordem 2 (RK2)**

* #### O método RK2 é uma técnica numérica utilizada para resolver equações diferenciais.
* #### Ele é especialmente útil para equações estocásticas, como a equação de Langevin.
* #### O método envolve a iteração de passos de tempo discretos para obter uma solução aproximada.

<br>

### **Aplicação do RK2 para a Equação de Langevin**

#### Para resolver a equação de Langevin com o método RK2, podemos seguir os seguintes passos:
1. #### Definir um passo de tempo $∆t$.
2. #### Calcular as forças aleatórias $η1(t)$ e $η2(t)$ no tempo atual.
3. #### Calcular as velocidades intermediárias $u′$ e $v′$ usando as forças no tempo atual.
4. #### Calcular as forças aleatórias $η1(t + ∆t)$ e $η2(t + ∆t)$ no próximo passo de tempo.
5. #### Atualizar as velocidades $u$ e $v$ para o próximo passo de tempo usando as velocidades intermediárias e as forças aleatórias no próximo passo de tempo.

<br>

### **Vantagens do RK2 para Equações Estocásticas**

#### O método RK2 oferece algumas vantagens ao resolver equações de Langevin estocásticas:
* #### É uma técnica simples e fácil de implementar.
* #### Possui uma precisão razoável para muitas aplicações.
* #### É mais estável e eficiente do que métodos de ordem inferior, como o método de Euler.
* #### Pode lidar com as forças aleatórias de forma adequada incorporando-as nos cálculos das velocidades intermediárias.

## **Solução da Equação de Langevin com o Método de Euler-Maruyama**


### **Método de Euler-Maruyama**

* #### O método de Euler-Maruyama é um método numérico comumente usado para resolver equações diferenciais estocásticas, como a equação de Langevin.
* #### Ele é uma extensão do método de Euler para incluir o termo estocástico.
* #### O método de Euler-Maruyama utiliza uma abordagem de passo de tempo discreto para obter uma solução aproximada.

<br>

### **Aplicação do Método de Euler-Maruyama para a Equação de Langevin**

#### Para resolver a equação de Langevin com o método de Euler-Maruyama,, podemos seguir os seguintes passos:
1. #### Definir um passo de tempo $∆t$.
2. #### Gerar uma amostra de ruído estocástico $∆W$, seguindo uma distribuição normal.
3. #### Calcular as forças aleatórias $η1(t)$ e $η2(t)$ no tempo atual.
4. #### Atualizar as velocidades $u$ e $v$ para o próximo passo de tempo usando as seguintes fórmulas: $$u(t + ∆t) = u(t) − γ_1u(t)∆t + F_1∆t +
\sqrt{∆t}η_1(t)$$ $$u(t + ∆t) = v(t) − γ_2v(t)∆t + F_2∆t +
\sqrt{∆t}η_2(t)$$

## **Filtro de Ornstein-Uhlenbeck (OU)**

* #### O processo estocástico de Ornstein-Uhlenbeck é um modelo amplamente utilizado para descrever a dinâmica de sistemas sujeitos a forças de atrito e ruído.
* #### O filtro de Ornstein-Uhlenbeck é uma técnica que permite filtrar um sinal, suavizando-o e removendo o ruído de alta frequência.
* #### O filtro de OU é baseado em uma equação diferencial estocástica chamada de Equação de Langevin.
* #### Essa equação descreve a taxa de mudança do sinal filtrado em relação ao tempo, levando em conta um termo de atrito e um termo de ruído gaussiano.

<br>

### **Ruído de Ornstein-Uhlenbeck (OU)**

#### O ruído de Ornstein-Uhlenbeck é um processo estocástico que pode ser descrito pela seguinte equação: $$dX_t = θ(μ − X_t)dt + σdW_t$$ onde:

* #### $X_t$ é o valor do processo no tempo $t$
* #### $θ$ é o coeficiente de retração em direção a $μ$
* #### $μ$ é o valor médio de longo prazo
* #### $σ$ é a magnitude do ruído aleatório
* #### $dW_t$ é o diferencial do movimento browniano (ruído branco)

<br>

#### Essa equação descreve a dinâmica do processo OU, onde o primeiro termo representa o retorno em direção a $μ$ e o segundo termo representa o ruído aleatório.

<br>

### **Outro Ruído Conhecido**

#### Existem vários tipos de ruídos conhecidos, um exemplo comum é o ruído branco, que pode ser descrito pela seguinte fórmula: $$dX_t = σdW_t$$ onde:
* #### $X_t$ é o valor do processo no tempo $t$
* #### $σ$ é a magnitude do ruído aleatório
* #### $dW_t$ é o diferencial do movimento browniano (ruído branco)

<br>

#### O ruído branco é caracterizado por ter uma distribuição de probabilidade constante e uma variância constante ao longo do tempo. E amplamente utilizado em an álise de sinais e em muitas outras aplicações.

# **Experimentos com os códigos**

O código abaixo realiza a simulação numérica da Equação de Langevin, que descreve a evolução temporal de uma partícula sujeita a uma força determinística e uma força estocástica, usando o método de Runge-Kutta de segunda ordem. O gráfico final exibe a posição da partícula em função do tempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the deterministic force u(t) and random force v(t) as synthetic functions
def u(t):
    return np.sin(2 * np.pi * t)  # Example: deterministic force as a sine function

def v(t):
    return np.random.normal(0, 1)  # Example: random force as a Gaussian distribution with mean 0 and standard deviation 1

# Function to compute the derivative dx/dt using the Langevin equation
def langevin_eq(t, x):
    return u(t) + v(t)

# Second-order Runge-Kutta (RK2) method for numerical integration
def rk2_method(t0, x0, dt, num_steps):
    t_values = [t0]
    x_values = [x0]

    for _ in range(num_steps):
        t = t_values[-1]
        x = x_values[-1]

        k1 = dt * langevin_eq(t, x)
        k2 = dt * langevin_eq(t + dt, x + k1)

        x_new = x + 0.5 * (k1 + k2)
        t_new = t + dt

        x_values.append(x_new)
        t_values.append(t_new)

    return t_values, x_values

# Define the time span and step size for the numerical integration
t0 = 0
t_end = 10
dt = 0.01
num_steps = int((t_end - t0) / dt)

# Set the initial condition x0 at time t0
x0 = 0.0

# Numerically solve the Langevin equation using the RK2 method
t_values, x_values = rk2_method(t0, x0, dt, num_steps)

# Plot the solution
plt.figure(figsize=(8, 6))
plt.plot(t_values, x_values, label='Numerical Solution')
plt.xlabel('Time (t)')
plt.ylabel('Position (x)')
plt.title('Langevin Equation Solution using RK2 Method')
plt.legend()
plt.grid(True)
plt.show()
```

O gráfico visualiza a evolução da posição de uma partícula ao longo do tempo, incorporando tanto componentes determinísticas quanto estocásticas, fornecendo uma representação numérica da solução da Equação de Langevin para o sistema em estudo. O caráter estocástico é evidenciado pela variabilidade na trajetória da partícula. Vejamos como o gráfico se comporta para outros parâmetros, forças e condições iniciais:

```python
def u(t):
    return np.sin(4 * np.pi * t)

def v(t):
    return 0.5 * np.random.normal(0, 1)

# Define the time span and step size for the numerical integration
t0 = 0
t_end = 10
dt = 0.1
num_steps = int((t_end - t0) / dt)

# Set the initial condition x0 at time t0
x0 = 2.0

# Numerically solve the Langevin equation using the RK2 method
t_values, x_values = rk2_method(t0, x0, dt, num_steps)

# Plot the solution
plt.figure(figsize=(8, 6))
plt.plot(t_values, x_values, label='Numerical Solution')
plt.xlabel('Time (t)')
plt.ylabel('Position (x)')
plt.title('Langevin Equation Solution using RK2 Method')
plt.legend()
plt.grid(True)
plt.show()
```

O gráfico gerado com esses novos parâmetros reflete as mudanças introduzidas nas forças determinísticas e estocásticas e na condição inicial. A frequência mais alta na força determinística se reflete em oscilações mais rápidas na posição da partícula ao longo do tempo. A redução na amplitude da força estocástica diminui a magnitude das flutuações aleatórias, tornando a trajetória menos volátil.

Em resumo, as alterações nos parâmetros fornecem insights sobre como diferentes aspectos do sistema afetam a evolução da posição da partícula, destacando a sensibilidade da solução numérica às condições iniciais e às características das forças determinísticas e estocásticas.

O código a seguir implementa a simulação de um sistema bidimensional usando as equações de Langevin para as componentes x e y. Ele obtém uma visualização das soluções das equações de Langevin para diferentes condições iniciais, ilustrando como as partículas se movem no plano x,y sob a influência de forças determinísticas e estocásticas. Os padrões e a forma da trajetória podem variar com base nas condições iniciais e nas características do ruído estocástico.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função para gerar ruído de Ornstein-Uhlenbeck
def generate_ou_noise(theta, mu, sigma, dt, num_steps):
    noise = np.zeros(num_steps)
    for i in range(1, num_steps):
        noise[i] = noise[i-1] + theta * (mu - noise[i-1]) * dt + sigma * np.sqrt(dt) * np.random.normal(0, 1)
    return noise

# Define as funções determinísticas u_x(t) e u_y(t) e as forças aleatórias v_x(t) e v_y(t) como funções sintéticas
def u_x(t):
    return np.sin(2 * np.pi * t)  # Exemplo: força determinística para a componente x como uma função seno

def u_y(t):
    return np.cos(2 * np.pi * t)  # Exemplo: força determinística para a componente y como uma função cosseno

# Função para calcular as derivadas dx/dt e dy/dt usando as equações de Langevin
def langevin_eq(t, x, y, noise_x):
    dx_dt = u_x(t) + noise_x  # Adiciona o ruído UO à componente x
    dy_dt = u_y(t) + np.random.normal(0, 1)  # Ruído branco para a componente y
    return dx_dt, dy_dt

# Método de Runge-Kutta de segunda ordem (RK2) para integração numérica
def rk2_method(t0, x0, y0, dt, num_steps):
    t_values = [t0]
    x_values = [x0]
    y_values = [y0]

    theta = 0.5
    mu = 0.0
    sigma = 1.0

    noise_x = generate_ou_noise(theta, mu, sigma, dt, num_steps + 1)  # Aumenta o tamanho do array de ruído para num_steps + 1

    for _ in range(num_steps):
        t = t_values[-1]
        x = x_values[-1]
        y = y_values[-1]

        k1_x, k1_y = dt * np.array(langevin_eq(t, x, y, noise_x[_]))
        k2_x, k2_y = dt * np.array(langevin_eq(t + dt, x + k1_x, y + k1_y, noise_x[_+1]))

        x_new = x + 0.5 * (k1_x + k2_x)
        y_new = y + 0.5 * (k1_y + k2_y)
        t_new = t + dt

        x_values.append(x_new)
        y_values.append(y_new)
        t_values.append(t_new)

    return t_values, x_values, y_values

# Defina o intervalo de tempo e o tamanho do passo para a integração numérica
t0 = 0
t_end = 20
dt = 0.01
num_steps = int((t_end - t0) / dt)

# Defina as condições iniciais x0 e y0 no tempo t0
x0 = [0, 1, -1]
y0 = [0, 1, -1]

# Execute o método de Runge-Kutta de segunda ordem para cada conjunto de condições iniciais
solutions = []
for i in range(len(x0)):
    t_values, x_values, y_values = rk2_method(t0, x0[i], y0[i], dt, num_steps)
    solutions.append((t_values, x_values, y_values))

# Plotar as soluções das componentes x e y para cada condição inicial
for i in range(len(solutions)):
    t_values, x_values, y_values = solutions[i]

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(t_values, x_values, label='x(t)')
    plt.plot(t_values, y_values, label='y(t)')
    plt.xlabel('Tempo (t)')
    plt.ylabel('Posição')
    plt.title('Soluções das Equações de Langevin (Componentes x e y)')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(x_values, y_values, label='Condição Inicial {}'.format(i + 1))
    plt.xlabel('Posição x')
    plt.ylabel('Posição y')
    plt.title('Trajetória no Espaço (x, y)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

Vemos que o ruído estocástico contribui para a variabilidade nas trajetórias, criando regiões mais densas e difusas. Comparando as trajetórias de diferentes conjuntos de condições iniciais, pode-se notar como pequenas variações nas condições iniciais resultam em comportamentos distintos. Portanto, esses gráficos fornecem uma compreensão visual da dinâmica do sistema sob a influência de forças determinísticas e estocásticas. Eles destacam a complexidade emergente de trajetórias que resultam da combinação de movimentos determinísticos e flutuações estocásticas. A interpretação desses gráficos ajuda a entender como diferentes fatores afetam a evolução de um sistema ao longo do tempo.

Este código fornece uma visão visual da dinâmica de um sistema bidimensional sujeito a forças determinísticas e estocásticas, contribuindo para uma compreensão mais profunda dos fenômenos complexos modelados pelas equações de Langevin.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define as funções determinísticas u_x(t) e u_y(t) e as forças aleatórias v_x(t) e v_y(t) como funções sintéticas
def u_x(t):
    return np.sin(2 * np.pi * t)  # Exemplo: força determinística para a componente x como uma função seno

def u_y(t):
    return np.cos(2 * np.pi * t)  # Exemplo: força determinística para a componente y como uma função cosseno

def v_x(t):
    return np.random.normal(0, 1)  # Exemplo: força aleatória para a componente x como uma distribuição gaussiana com média 0 e desvio padrão 1

def v_y(t):
    return np.random.normal(0, 1)  # Exemplo: força aleatória para a componente y como uma distribuição gaussiana com média 0 e desvio padrão 1

# Função para calcular as derivadas dx/dt e dy/dt usando as equações de Langevin
def langevin_eq(t, x, y):
    dx_dt = u_x(t) + v_x(t)
    dy_dt = u_y(t) + v_y(t)
    return dx_dt, dy_dt

# Método de Runge-Kutta de segunda ordem (RK2) para integração numérica
def rk2_method(t0, x0, y0, dt, num_steps):
    t_values = [t0]
    x_values = [x0]
    y_values = [y0]

    for _ in range(num_steps):
        t = t_values[-1]
        x = x_values[-1]
        y = y_values[-1]

        k1_x, k1_y = dt * np.array(langevin_eq(t, x, y))
        k2_x, k2_y = dt * np.array(langevin_eq(t + dt, x + k1_x, y + k1_y))

        x_new = x + 0.5 * (k1_x + k2_x)
        y_new = y + 0.5 * (k1_y + k2_y)
        t_new = t + dt

        x_values.append(x_new)
        y_values.append(y_new)
        t_values.append(t_new)

    return t_values, x_values, y_values

# Defina o intervalo de tempo e o tamanho do passo para a integração numérica
t0 = 0
t_end = 160
dt = 0.01
num_steps = int((t_end - t0) / dt)

# Defina as condições iniciais x0 e y0 no tempo t0
x0 = [0.0, 2.0, -1.0]
y0 = [1.0, -0.5, 3.0]

# Resolva numericamente as equações de Langevin para cada conjunto de condições iniciais
solutions = []
for i in range(len(x0)):
    t_values, x_values, y_values = rk2_method(t0, x0[i], y0[i], dt, num_steps)
    solutions.append((t_values, x_values, y_values))

# Plotar as soluções das componentes x e y com cores diferentes para cada condição inicial
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for i in range(len(solutions)):
    t_values, x_values, y_values = solutions[i]
    plt.plot(t_values, x_values, label='x{}(t)'.format(i + 1))
    plt.plot(t_values, y_values, label='y{}(t)'.format(i + 1))
plt.xlabel('Tempo (t)')
plt.ylabel('Posição')
plt.title('Soluções das Equações de Langevin (Componentes x e y)')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
for i in range(len(solutions)):
    x_values, y_values = solutions[i][1:]
    plt.plot(x_values, y_values, label='Condição Inicial {}'.format(i + 1))
plt.xlabel('Posição x')
plt.ylabel('Posição y')
plt.title('Trajetória no Espaço (x, y)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Este conjunto de gráficos oferece uma visão abrangente da dinâmica do sistema bidimensional sob a influência de forças determinísticas e estocásticas. As diferentes cores representam diferentes conjuntos de condições iniciais, evidenciando como pequenas variações nas condições iniciais resultam em trajetórias distintas. As oscilações regulares são influenciadas pelas forças determinísticas, enquanto as flutuações irregulares são causadas pelas forças estocásticas.
```python
# Defina o intervalo de tempo e o tamanho do passo para a integração numérica
t0 = 0
t_end = 50
dt = 0.02
num_steps = int((t_end - t0) / dt)

# Defina as condições iniciais x0 e y0 no tempo t0
x0 = [1.0, -2.0, 0.5]
y0 = [-0.5, 1.5, -2.0]

# Plotar as soluções das componentes x e y com cores diferentes para cada condição inicial
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for i in range(len(solutions)):
    t_values, x_values, y_values = solutions[i]
    plt.plot(t_values, x_values, label='x{}(t)'.format(i + 1))
    plt.plot(t_values, y_values, label='y{}(t)'.format(i + 1))
plt.xlabel('Tempo (t)')
plt.ylabel('Posição')
plt.title('Soluções das Equações de Langevin (Componentes x e y)')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
for i in range(len(solutions)):
    x_values, y_values = solutions[i][1:]
    plt.plot(x_values, y_values, label='Condição Inicial {}'.format(i + 1))
plt.xlabel('Posição x')
plt.ylabel('Posição y')
plt.title('Trajetória no Espaço (x, y)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

O código a seguir implementa um modelo de Langevin para simular o movimento de partículas em um espaço bidimensional. Ele oferece uma representação gráfica das trajetórias de partículas sujeitas a forças determinísticas e estocásticas, proporcionando insights sobre o comportamento do sistema ao longo do tempo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define as funções determinísticas u_x(t) e u_y(t) e as forças aleatórias v_x(t) e v_y(t) como funções sintéticas
def u_x(t):
    return np.sin(2 * np.pi * t)  # Exemplo: força determinística para a componente x como uma função seno

def u_y(t):
    return np.cos(2 * np.pi * t)  # Exemplo: força determinística para a componente y como uma função cosseno

def v_x(t, x):
    theta = 0.5
    mu = 0.0
    sigma = 1.0
    return theta * (mu - x) + sigma * np.random.normal(0, 1)  # Ruído de Ornstein-Uhlenbeck para a componente x

def v_y(t):
    return np.random.normal(0, 1)  # Ruído branco para a componente y

# Função para calcular as derivadas dx/dt e dy/dt usando as equações de Langevin
def langevin_eq(t, x, y):
    dx_dt = u_x(t) + v_x(t, x)
    dy_dt = u_y(t) + v_y(t)
    return dx_dt, dy_dt

# Método de Runge-Kutta de segunda ordem (RK2) para integração numérica
def rk2_method(t0, x0, y0, dt, num_steps):
    t_values = [t0]
    x_values = [x0]
    y_values = [y0]

    for _ in range(num_steps):
        t = t_values[-1]
        x = x_values[-1]
        y = y_values[-1]

        k1_x, k1_y = dt * np.array(langevin_eq(t, x, y))
        k2_x, k2_y = dt * np.array(langevin_eq(t + dt, x + k1_x, y + k1_y))

        x_new = x + 0.5 * (k1_x + k2_x)
        y_new = y + 0.5 * (k1_y + k2_y)
        t_new = t + dt

        x_values.append(x_new)
        y_values.append(y_new)
        t_values.append(t_new)

    return t_values, x_values, y_values

# Defina o intervalo de tempo e o tamanho do passo para a integração numérica
t0 = 0
t_end = 10
dt = 0.01
num_steps = int((t_end - t0) / dt)

# Defina as condições iniciais x0 e y0 no tempo t0
x0 = [0.0, 2.0, -1.0]
y0 = [1.0, -0.5, 3.0]

# Resolva numericamente as equações de Langevin para cada conjunto de condições iniciais
solutions = []
for i in range(len(x0)):
    t_values, x_values, y_values = rk2_method(t0, x0[i], y0[i], dt, num_steps)
    solutions.append((t_values, x_values, y_values))

# Plotar as soluções das componentes x e y para cada condição inicial
for i in range(len(solutions)):
    t_values, x_values, y_values = solutions[i]

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(t_values, x_values, label='x(t)')
    plt.plot(t_values, y_values, label='y(t)')
    plt.xlabel('Tempo (t)')
    plt.ylabel('Posição')
    plt.title('Soluções das Equações de Langevin (Componentes x e y)')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(x_values, y_values, label='Condição Inicial {}'.format(i + 1))
    plt.xlabel('Posição x')
    plt.ylabel('Posição y')
    plt.title('Trajetória no Espaço (x, y)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

Este conjunto de gráficos oferece uma visão abrangente da dinâmica do sistema bidimensional sob a influência de forças determinísticas e estocásticas. A variação nas condições iniciais e a combinação de forças determinísticas e estocásticas contribuem para a complexidade das trajetórias observadas no espaço (x,y).
