# **Método de Monte Carlo**

> #### O método de Monte Carlo é uma técnica estatística e computacional amplamente utilizada para estimar resultados complexos ou calcular integrais, simulando uma grande quantidade de amostras aleatórias de um sistema ou processo. Ele é frequentemente aplicado em situações em que a solução analítica exata é difícil ou impossível de obter.

<br>

#### O nome "Monte Carlo" é uma referência ao famoso cassino localizado no Principado de Mônaco, conhecido por jogos de azar e aleatoriedade. O método foi batizado assim devido à natureza estocástica das simulações e à semelhança com jogos de azar.

<br>

#### A ideia principal por trás do método de Monte Carlo é utilizar números aleatórios para representar incertezas e variações em um problema e, em seguida, realizar cálculos com base nessas amostras aleatórias para estimar uma resposta. Aqui estão os passos gerais envolvidos:
<br>

1. #### Definição do problema: Primeiro, você deve formular o problema que deseja resolver em termos de uma função matemática, integral, ou sistema que envolva aleatoriedade ou incerteza.
2. #### Geração de números aleatórios: Gere uma grande quantidade de números aleatórios de acordo com uma distribuição apropriada. Esses números representam os parâmetros incertos ou variáveis do sistema.
3. #### Avaliação da função: Utilize os números aleatórios gerados para avaliar a função ou as equações do problema. Isso é feito para cada amostra aleatória.
4. #### Média e estatísticas: Calcule a média ou outras estatísticas das saídas obtidas a partir da avaliação da função. Essas estatísticas fornecem estimativas do valor desejado ou da solução do problema.
5. #### Aumento da precisão: À medida que mais amostras aleatórias são geradas, as estimativas tendem a convergir para o valor verdadeiro, aumentando a precisão da solução.

<br>

#### O método de Monte Carlo é especialmente útil em problemas multidimensionais e complexos, como simulações de sistemas físicos, previsão de preços de opções financeiras, otimização de processos industriais, análise de risco e muitos outros domínios. É uma técnica versátil que pode ser aplicada a uma ampla gama de problemas, desde que a geração de números aleatórios seja feita de maneira apropriada e as amostras sejam suficientemente numerosas para fornecer estimativas precisas.



## **Importância da Variância para o Amostragem de Dados**

- #### A variância é uma medida de dispersão que indica o quão distantes os valores de uma variável estão da média.
- #### Na amostragem de dados, a variância desempenha um papel fundamental.
- #### Quanto maior a variância, mais espalhados estão os dados em relação à média, o que pode afetar a precisão das estimativas obtidas a partir da amostra.
- #### Ao planejar uma amostragem, é importante considerar a variância da população-alvo e ajustar o tamanho da amostra de acordo para obter resultados mais confiáveis.

## **Lei dos Grandes Números**

- #### A Lei dos Grandes Números é um princípio fundamental da teoria das probabilidades e estatística.
- #### Ela afirma que, à medida que o tamanho da amostra aumenta, a média dos valores observados se aproxima cada vez mais da média populacional verdadeira.
- #### Em outras palavras, à medida que coletamos mais dados, a incerteza diminui e a estimativa da média torna-se mais precisa.
- #### A Lei dos Grandes Números é essencial para o método de Monte Carlo, pois fornece uma base teórica para a utilização de amostras aleatórias para estimar parâmetros populacionais.

## **Significado do Método de Monte Carlo**

- #### O método de Monte Carlo é uma poderosa ferramenta estatística que utiliza amostragem aleatória para estimar resultados ou calcular soluções aproximadas.
- #### Ele se baseia na ideia de que, ao gerar uma grande quantidade de números aleatórios relevantes para o problema em questão, é possível obter uma estimativa aproximada do resultado desejado.
- #### Ao repetir as simulações várias vezes, é possível reduzir a incerteza e melhorar a precisão da estimativa.
- #### O método de Monte Carlo é amplamente aplicado em várias áreas, como física, matemática, finanças, engenharia, ciência da computação, entre outras, para resolver problemas complexos e obter respostas aproximadas úteis.

### **Moeda**

O código abaixo simula uma série de lançamentos de uma moeda, onde cada lançamento pode resultar em "cara" (Heads) ou "coroa" (Tails). O código é uma simulação simples de lançamentos de moeda e é usado para contar quantas vezes "cara" e "coroa" aparecem em um número específico de lançamentos simulados. Isso demonstra como você podemos usar o módulo random em Python para realizar simulações estocásticas e contar resultados.

```python
import random

def simulate_coin_toss(num_tosses):
    heads = 0
    tails = 0

    for _ in range(num_tosses):
        toss = random.choice(['Heads', 'Tails'])

        if toss == 'Heads':
            heads += 1
        else:
            tails += 1

    return heads, tails

# Exemplo de uso
num_tosses = 10000
num_heads, num_tails = simulate_coin_toss(num_tosses)
print(f"Número de lançamentos de cara (Heads): {num_heads}")
print(f"Número de lançamentos de coroa (Tails): {num_tails}")
```

```python
import matplotlib.pyplot as plt

plt.bar(['Cara', 'Coroa'], [num_heads, num_tails])
plt.xlabel('Resultado do lançamento')
plt.ylabel('Quantidade de vezes')
plt.title(f'Resultados obtidos em {num_tosses} lançamentos')
plt.show()
```

Fizemos $10.000$ lançamentos e sabemos que temos a mesma chance de obter tanto cara como coroa $(50\%)$. Por isso, notamos que o número de lançamentos de cara tende a ser bem próximo do número de lançamentos de coroa, e esses dois valores, em qualquer execução, não devem estar muito longe de $5000$.

### **Roleta**

O código abaixo simula um giro em uma roleta, onde a roleta tem números de 0 a 36, sendo que o número 0 é colorido de verde, e os outros números são coloridos de preto se ele for par ou de vermelho se for ímpar. Em resumo, o código simula um giro em uma roleta, gerando aleatoriamente um número entre 0 e 36 e determinando a cor associada a esse número. Isso é útil para simular jogos de roleta em um programa e realizar análises ou experimentações relacionadas a jogos de azar.

```python
import random

def spin_roulette():
    numbers = list(range(0, 37))  # Números de 0 a 36 na roleta
    colors = ['black', 'red', 'green']  # Cores dos números na roleta

    winning_number = random.choice(numbers)  # Número vencedor aleatório
    winning_color = colors[2] if winning_number == 0 else colors[winning_number % 2]  # Cor do número vencedor

    return winning_number, winning_color

# Exemplo de uso
winning_number, winning_color = spin_roulette()

print(f"Número vencedor: {winning_number}")
print(f"Cor do número vencedor: {winning_color}")
```

Para essa execução, tiramos o número 32 que é par e, por isso, tem a cor preta. Vamos utilizar essa função da roleta para contar as quantidades de cada cor que tiramos em $10.000$ giros e mostrar isso graficamente.

```python
import matplotlib.pyplot as plt

num_giros = 10000
black, red, green = 0, 0, 0

for _ in range(num_giros):
  winning_number, winning_color = spin_roulette()

  if winning_color == 'black': black += 1
  elif winning_color == 'red': red += 1
  else: green += 1

barlist = plt.bar(['Preto', 'Vermelho', 'Verde'], [black, red, green])
barlist[0].set_color('k')
barlist[1].set_color('r')
barlist[2].set_color('g')
plt.xlabel('\nResultado do giro')
plt.ylabel('Quantidade de vezes')
plt.title(f'Resultados obtidos em {num_giros} giros')
plt.show()
```

Como de $1$ a $36$ temos a mesma quantidade de números pares e ímpares, a chance de obtermos um número preto é igual a de obtermos um número vermelho. Por isso, essas barras no gráfico apresentam um tamanho muito parecido. Enquanto a barra dos número verdes é bem menor, uma vez que temos a chance de $1/37$ de tirarmos o zero na roleta.

### **Amostras**

O código que abaixo implementa uma simulação de Monte Carlo para estimar o número de amostras necessárias para atingir uma determinada precisão (erro relativo) e nível de confiança em uma estimativa estatística. A simulação utiliza amostras aleatórias para estimar a média e o desvio padrão de uma distribuição e, em seguida, calcula o número de amostras necessárias para atender aos critérios de precisão e nível de confiança especificados.

```python
import random
import math

def monte_carlo_simulation(precision, confidence):
    num_samples = 100  # Número inicial de amostras
    mean = 0.0  # Média das amostras
    squared_deviations = []  # Lista para armazenar os desvios quadrados

    while True:
        samples = [random.random() for _ in range(num_samples)]
        mean = sum(samples) / num_samples
        squared_deviations = [(x - mean) ** 2 for x in samples]
        variance = sum(squared_deviations) / num_samples
        standard_deviation = math.sqrt(variance)
        coefficient_of_variation = standard_deviation / mean
        margin_of_error = 1.96 * (coefficient_of_variation / math.sqrt(num_samples))  # Intervalo de confiança de 95%
        if margin_of_error <= precision * mean:
            break
        num_samples *= 2  # Dobra o número de amostras

    return num_samples

# Exemplo de uso
precision = 0.01  # Precisão desejada
confidence = 0.95  # Nível de confiança

num_samples_needed = monte_carlo_simulation(precision, confidence)
print(f"Número de amostras necessárias: {num_samples_needed}")
```

Este código ilustra como o método de Monte Carlo pode ser usado para determinar o tamanho de amostra necessário para atender a critérios de precisão e confiança em uma estimativa estatística, como a média de uma distribuição. Ele faz isso iterativamente, aumentando o número de amostras até que a precisão desejada seja alcançada.

Estamos utilizando um valor de precisão de $0.01$, uma confiança de $0.95$ e começando sempre com um número inicial de $100$ amostras que dobra a cada iteração. Como as amostras são geradas de forma aleatória em cada vez, em cada execução podemos obter ao final um resultado diferente no número de amostras necessárias. Vamos rodar esse código $30$ vezes para observar se há repetições nos resultados.

```python
import matplotlib.pyplot as plt
from collections import Counter

precision = 0.01  # Precisão desejada
confidence = 0.95  # Nível de confiança
num_simulacoes = 30
results, qtds = [], []

for _ in range(num_simulacoes):
  num_samples = monte_carlo_simulation(precision, confidence)
  results.append(num_samples)

plt.plot(range(len(results)), results)
plt.grid(True)
```

Vemos que os resultados encontrados variam sempre entre dois valores: $51200$ e $102400$. Vejamos quai resultados teremos se utilizarmos outros valores de precisão e confiança.

```python
# Primeira combinação de precisão e confiança
precision1 = 0.01
confidence1 = 0.95
num_samples_needed1 = monte_carlo_simulation(precision1, confidence1)
print(f"Número de amostras necessárias (Precisão 1): {num_samples_needed1}")

# Segunda combinação de precisão e confiança
precision2 = 0.005
confidence2 = 0.99
num_samples_needed2 = monte_carlo_simulation(precision2, confidence2)
print(f"Número de amostras necessárias (Precisão 2): {num_samples_needed2}")

# Terceira combinação de precisão e confiança
precision3 = 0.001
confidence3 = 0.90
num_samples_needed3 = monte_carlo_simulation(precision3, confidence3)
print(f"Número de amostras necessárias (Precisão 3): {num_samples_needed3}")
```

Neste código, eu defini três combinações diferentes de valores para "precision" e "confidence" e, em seguida, calculei o número de amostras necessárias para cada combinação. Os resultados são impressos na tela para cada combinação. Vamos analisar esses valores graficamente a seguir.

```python
resultados = [num_samples_needed1, num_samples_needed2, num_samples_needed3]

plt.bar([1, 2, 3], resultados)
plt.xlabel('Combinação (Precisão e confiança)')
plt.ylabel('Número de amostras necessárias')
plt.title(f'Número de amostras necessárias para cada combinação')
plt.show()
```

Podemos ver que, para a combinação $3$, o número de amostras necessárias é muito maior que para as outras. Isso ocorre pois a combinação $3$ possui o menor valor de precisão $(0.001)$, assim como o menor nível de confiança $(0.90)$ entre as demais.

### **Grafo**

O código abaixo gera e visualiza um grafo aleatório usando a biblioteca NetworkX e o Matplotlib em Python. Ele gera um grafo aleatório usando o modelo de Erdős-Rényi com o número de nós e a probabilidade de aresta especificados. Em seguida, ele utiliza o Matplotlib para visualizar o grafo gerado, com rótulos nos nós, cores personalizadas e outros atributos de estilo. Isso é útil para visualizar e estudar grafos aleatórios ou outros tipos de redes.

```python
import networkx as nx
import matplotlib.pyplot as plt

def generate_random_graph(num_nodes, probability):
    graph = nx.erdos_renyi_graph(num_nodes, probability)
    return graph

# Exemplo de uso
num_nodes = 10  # Número de nós no grafo
probability = 0.4  # Probabilidade de existir uma aresta entre dois nós

# Gerar o grafo aleatório
graph = generate_random_graph(num_nodes, probability)

# Plotar o gráfico resultante
nx.draw(graph, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')
plt.title("Grafo Aleatório")
plt.show()
```

### **Física**

O código a seguir simula o decaimento radioativo de um conjunto de núcleos atômicos ao longo do tempo. O decaimento radioativo é um processo estocástico no qual os núcleos podem decair com uma certa probabilidade em um determinado intervalo de tempo. Ele simula o decaimento radioativo de um conjunto de núcleos iniciais, onde cada núcleo tem uma probabilidade de decair durante um intervalo de tempo. A fração de núcleos que decaiu é calculada e impressa como resultado da simulação. Isso é útil para modelar processos de decaimento radioativo em experimentos ou estudos científicos.

```python
import random

def simulate_radioactive_decay(num_nuclei, decay_probability):
    num_decayed = 0

    for _ in range(num_nuclei):
        if random.random() < decay_probability:
            num_decayed += 1

    decay_fraction = num_decayed / num_nuclei
    return decay_fraction

# Exemplo de uso
num_nuclei = 1000  # Número de núcleos iniciais
decay_probability = 0.02  # Probabilidade de decaimento de um núcleo

decay_fraction = simulate_radioactive_decay(num_nuclei, decay_probability)

print(f"Fração de núcleos decaídos: {decay_fraction}")
```

### **Distribuição**

O código abaixo gera uma amostra de números aleatórios que seguem uma distribuição normal com a média e o desvio padrão especificados. Isso é útil em diversas aplicações, como simulações estatísticas, modelagem de fenômenos naturais e experimentos em que as observações são esperadas para seguir uma distribuição normal. A biblioteca NumPy é amplamente utilizada para realizar operações matemáticas e estatísticas em Python.

```python
import numpy as np

def generate_normal_random_numbers(mean, std, size):
    return np.random.normal(mean, std, size)

mean = 0  # média da distribuição normal
std = 1  # desvio padrão da distribuição normal
size = 30  # número de amostras a serem geradas

random_numbers = generate_normal_random_numbers(mean, std, size)
print(random_numbers)
```

Essas foram as $30$ amostras que foram geradas de forma aleatória seguindo uma distribuição normal. Vejamos como se apresenta o histograma para uma quantidade de $10000$ amostras:

```python
import matplotlib.pyplot as plt

mean = 0  # média da distribuição normal
std = 1  # desvio padrão da distribuição normal
size = 10000  # número de amostras a serem geradas

random_numbers = generate_normal_random_numbers(mean, std, size)

plt.title('Histograma', fontsize=15)
plt.xlabel('Número aleatório', fontsize=10)
plt.ylabel('Frequência Absoluta', fontsize=10)
plt.tick_params(labelsize=15)
plt.hist(random_numbers, 100, rwidth=0.9, color='c', alpha=0.7)
plt.show()
```

Podemos perceber, ao observar o histograma, que os números das amostram realmente seguem uma distribuição normal. Faremos o mesmo processo, porém agora gerando números aleatórios a partir de uma distribuição de Pareto.

```python
import numpy as np

def generate_pareto_random_numbers(shape, size):
    return np.random.pareto(shape, size)

shape = 2.5  # parâmetro de forma da distribuição de Pareto
size = 30  # número de amostras a serem geradas

random_numbers = generate_pareto_random_numbers(shape, size)
print(random_numbers)
```

```python
import matplotlib.pyplot as plt

shape = 2.5  # parâmetro de forma da distribuição de Pareto
size = 500  # número de amostras a serem geradas

random_numbers = generate_pareto_random_numbers(shape, size)

plt.title('Histograma', fontsize=15)
plt.xlabel('Número aleatório', fontsize=10)
plt.ylabel('Frequência Absoluta', fontsize=10)
plt.tick_params(labelsize=15)
plt.hist(random_numbers, 50, rwidth=0.9, color='r', alpha=0.7, edgecolor='black')
plt.show()
```

Em resumo, o código gera uma amostra de números aleatórios que segue uma distribuição de Pareto com o parâmetro de forma especificado. A distribuição de Pareto é caracterizada por sua cauda longa e é frequentemente usada para modelar eventos raros com alta variabilidade. Ela é aplicada em situações em que a maioria dos eventos tem baixa probabilidade de ocorrência, mas alguns poucos eventos têm probabilidade significativamente maior.
