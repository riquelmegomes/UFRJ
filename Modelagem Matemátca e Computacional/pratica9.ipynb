# **Modelo SIR (Suscetível-Infectado-Recuperado)**

> #### O modelo SIR é um dos modelos epidemiológicos mais clássicos e amplamente utilizados para descrever a propagação de doenças infecciosas em uma população. Nesse modelo, a população é dividida em três compartimentos:
- #### **Suscetíveis $(S)$:** indivíduos que podem ser infectados pela doença.
- #### **Infectados $(I)$:** indivíduos atualmente infectados e capazes de transmitir a doença.
- #### **Recuperados $(R)$:** indivíduos que se recuperaram da doença e adquiriram imunidade.

<br>

#### O modelo SIR é usado para entender a propagação de doenças, prever a evolução de uma epidemia e avaliar a eficácia das medidas de controle, como vacinação ou quarentena. Ele é um modelo simplificado e não leva em consideração fatores complexos, como taxas de natalidade, morte natural e migração, mas fornece uma estrutura básica para analisar epidemias infecciosas.

## **Equações diferenciais do modelo SIR**

#### O modelo SIR é descrito por um sistema de equações diferenciais que descrevem as taxas de mudança dos compartimentos ao longo do tempo. As equações diferenciais do modelo SIR são as seguintes:
#### $$\frac{dS}{dt}=-\frac{β·S⋅I}{N}$$
#### $$\frac{dI}{dt}=\frac{β·S⋅I}{N}-γ⋅I$$
#### $$\frac{dR}{dt}=γ⋅I$$
#### onde,
- #### $S:$ número de indivíduos suscetíveis à doença,
- #### $I:$ número de indivíduos infectados,
- #### $R:$ número de indivíduos recuperados,
- #### $N:$ população total $(N = S + I + R)$,
- #### $β:$ taxa de transmissão da doença,
- #### $γ:$ taxa de recuperação da doença.


### **Interpretação dos parâmetros do modelo SIR**

#### Os parâmetros do modelo SIR têm interpretações importantes na compreensão da dinâmica da doença:
- #### **$β:$** taxa de transmissão da doença. Representa a probabilidade de um indivíduo suscetível se tornar infectado ao entrar em contato com um indivíduo infectado. Quanto maior o valor de $β$, maior a taxa de propagação da doença.
- #### **$γ:$** taxa de recuperação da doença. Representa a taxa na qual os indivíduos infectados se recuperam e adquirem imunidade. Quanto maior o valor de $γ$, mais rápido os indivíduos se recuperam da doença.
- #### **$R_0:$** número básico de reprodução. É definido como a média de novas infecções causadas por um indivíduo infectado em uma população totalmente suscetvel. É calculado como $R_0=\frac{β}{γ}$. Se $R_0>1$, a doença tem potencial de se espalhar em uma população suscetível.

<br>

#### Além dos parâmetros, existem outras quantidades importantes derivadas do modelo SIR:

- #### **$D:$** duração média da doença. É calculada como $D=\frac{1}{γ}$ e representa o tempo médio que um indivíduo infectado permanece na fase infecciosa antes de se recuperar.
- #### **$\frac{1}{γ}:$** período infeccioso médio. É o inverso da taxa de recuperação e representa o tempo médio que um indivíduo infectado permanece infeccioso antes de se recuperar.
<br>

#### Essas quantidades ajudam a entender a dinâmica temporal da doença e a avaliar o impacto das intervenções, como isolamento social e vacinação.

## **Generalização do modelo SIR**

#### O modelo SIR pode ser generalizado e estendido para levar em consideração diferentes características das epidemias. Algumas das extensões mais comuns são:
- #### **Modelo SIRD**: inclui uma quarta classe de compartimento $(D)$ que representa os indivíduos que morreram devido à doença.
- #### **Modelo SEIR**: adiciona um compartimento adicional $(E)$ que representa os indivíduos expostos à doença, mas ainda não infectados.
- #### **Modelo SEIRD**: combina as extensões do modelo SEIR e SIRD levando em consideração a exposição e a mortalidade.
- #### Outros modelos: existem muitas outras variações e extensões do modelo SIR, como modelos que consideram diferentes faixas etárias, modelos que incorporam intervenções ou modelos espaciais que levam em conta a distribuição geográfica da população.

<br>

#### Essas extensões permitem uma representação mais precisa e realista das características das epidemias.

### **Modelo SIRD**

#### O modelo SIRD adiciona um compartimento adicional para representar os indivíduos que morreram devido à doença. As equações diferenciais para o modelo SIRD são as seguintes:

#### $$\frac{dS}{dt}=-\frac{β·S⋅I}{N}$$
#### $$\frac{dI}{dt}=\frac{β·S⋅I}{N}-γ⋅I-μ⋅I$$
#### $$\frac{dR}{dt}=γ⋅I$$
#### $$\frac{dD}{dt}=μ⋅I$$
#### onde,
- $S:$ número de indivíduos suscetíveis à doença,
- $I:$ número de indivíduos infectados,
- $R:$ número de indivíduos recuperados,
- $D:$ número de indivíduos que morreram,
- $N:$ população total $(N = S + I + R + D)$,
- $β:$ taxa de transmissão da doença,
- $γ:$ taxa de recuperação da doença,
- $μ:$ taxa de mortalidade da doença.

### **Modelo SEIR**

#### O modelo SEIR adiciona um compartimento adicional para representar os indivíduos expostos à doença. As equações diferenciais para o modelo SEIR são as seguintes:

#### $$\frac{dS}{dt}=-\frac{β·S⋅I}{N}$$
#### $$\frac{dE}{dt}=\frac{β·S⋅I}{N}-\alpha⋅E$$
#### $$\frac{dI}{dt}=\alpha⋅E-γ⋅I$$
#### $$\frac{dR}{dt}=γ⋅I$$
onde,
- $S:$ número de indivíduos suscetíveis à doença,
- $E:$ número de indivíduos expostos à doença,
- $I:$ número de indivíduos infectados,
- $R:$ número de indivíduos recuperados,
- $N:$ população total $(N = S + E + I + R)$,
- $β:$ taxa de transmissão da doença,
- $γ:$ taxa de recuperação da doença,
- $α:$ taxa de exposição à doença.


### **Modelo SEIRD**

#### O modelo SEIRD combina as extensões do modelo SEIR e SIRD. As equações diferenciais para o modelo SEIRD são as seguintes:

#### $$\frac{dS}{dt}=-\frac{β·S⋅I}{N}$$
#### $$\frac{dE}{dt}=\frac{β·S⋅I}{N}-\alpha⋅E$$
#### $$\frac{dI}{dt}=\alpha⋅E-γ⋅I-μ⋅I$$
#### $$\frac{dR}{dt}=γ⋅I$$
#### $$\frac{dD}{dt}=μ⋅I$$
onde,
- $S:$ número de indivíduos suscetíveis à doença,
- $E:$ número de indivíduos expostos à doença,
- $I:$ número de indivíduos infectados,
- $R:$ número de indivíduos recuperados,
- $D:$ número de indivíduos que morreram,
- $N:$ população total $(N = S + E + I + R + D)$,
- $β:$ taxa de transmissão da doença,
- $γ:$ taxa de recuperação da doença,
- $α:$ taxa de exposição à doença.
- $μ:$ taxa de mortalidade da doença.



### **Utilidade das generalizações**

#### As generalizações do modelo SIR são úteis para representar diferentes características das epidemias, como:
- #### Tempo de incubação: o modelo SEIR considera a fase de exposição, onde os indivíduos estão infectados, mas ainda não são infecciosos.
- #### Mortalidade: o modelo SIRD leva em conta a taxa de mortalidade da doença, permitindo avaliar o impacto dos óbitos na população.
- #### Intervenções e políticas de saúde pública: as generalizações do modelo SIR podem incorporar parâmetros que representam medidas de controle, como vacinação, isolamento social, uso de máscaras, entre outros.
- #### Compartimentalização da população: os modelos baseados em compartimentos permitem segmentar a população em grupos específicos, considerando fatores como faixa etária, localização geográfica, comorbidades, entre outros.

<br>

#### Essas generalizações fornecem uma maior flexibilidade e precisão para analisar e prever o comportamento das epidemias.

### **SIR**

O código define um modelo SIR, simula a propagação de uma doença infecciosa e gera um gráfico que mostra como as populações de suscetíveis, infectados e recuperados evoluem ao longo do tempo, além de calcular e imprimir o valor de R0, que é a razão entre a taxa de transmissão (beta) e a taxa de recuperação (gamma). A função solve_ivp é usada para resolver as equações diferenciais do modelo SIR. Ela recebe como argumentos a função sir_model, o intervalo de tempo [t0, tf], as condições iniciais [S0, I0, R0], e o vetor t_eval que especifica os pontos de tempo em que a solução será avaliada.

```python
import numpy as np
from scipy.integrate import solve_ivp, odeint
import matplotlib.pyplot as plt

# Função que define o sistema de equações diferenciais do modelo SIR
def sir_model(t, y, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]

# Parâmetros do modelo SIR
beta = 0.4  # Taxa de transmissão
gamma = 0.1  # Taxa de recuperação

# Condições iniciais
S0 = 0.99  # População suscetível inicial
I0 = 0.01  # População infectada inicial
R0 = 0.0  # População recuperada inicial

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 100.0  # Tempo final
dt = 0.1  # Passo de integração

# Resolvendo o sistema de equações diferenciais
sol = solve_ivp(lambda t, y: sir_model(t, y, beta, gamma), [t0, tf], [S0, I0, R0], t_eval=np.arange(t0, tf, dt))
#sol = odeint(lambda t, y: sir_model(t, y, beta, gamma), [t0, tf], [S0, I0, R0], t_eval=np.arange(t0, tf, dt))

# Plotando o gráfico da solução
plt.plot(sol.t, sol.y[0], label='Suscetíveis')
plt.plot(sol.t, sol.y[1], label='Infectados')
plt.plot(sol.t, sol.y[2], label='Recuperados')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SIR')
plt.grid(True)
plt.legend()
plt.show()

print(f"\nR0 = {beta/gamma}")
```

A curva em azul representa a população suscetível $(S)$ ao longo do tempo. Inicialmente, a maior parte da população é suscetível, mas à medida que o tempo avança, essa população diminui devido à transmissão da doença.

A curva em laranja representa a população infectada $(I)$. Ela começa em um valor baixo $(I0)$ e aumenta gradualmente à medida que a doença se propaga. A curva atinge um pico e começa a diminuir quando a taxa de recuperação (gamma) é efetiva.

A curva em verde representa a população recuperada $(R)$. Inicialmente, a população recuperada é zero, mas aumenta à medida que mais pessoas se recuperam da doença.

Vejamos como o gráfico se apresenta para diferentes parâmetros e condições iniciais.

Nesse último gráfico, a população suscetível atinge um valor mais baixo e a curva da população recuperada também cresce mais rapidamente devido à maior taxa de recuperação. Além disso, o pico da curva de infectados $(I)$ ocorre mais cedo no código modificado devido à menor taxa de transmissão, resultando em um pico de infecção mais baixo em comparação com o código original.

### **Métodos numéricos**

O código a seguir implementa três funções que são usadas para resolver equações diferenciais ordinárias (EDOs) usando diferentes métodos numéricos: Euler melhorados, Runge-Kutta de ordem 3 e Runge-Kutta de ordem 4. As três funções retornam os valores de tempo (t_values) e as soluções numéricas das EDOs em um formato que permite a fácil representação dos resultados.

```python
# Função que implementa o método de Euler melhorado
def euler_melhorado(f, t0, tf, y0, h):
    t_values = np.arange(t0, tf, h)
    y_values = [y0]

    for t in t_values[:-1]:
        y_n = y_values[-1]
        k1 = np.array(f(t, y_n))
        k2 = np.array(f(t + h, y_n + h*k1))
        y_n1 = y_n + 0.5 * h * (k1 + k2)
        y_values.append(y_n1)

    return t_values, np.array(y_values).T

# Função que implementa o método de Runge-Kutta de ordem 3
def runge_kutta_ordem_3(f, t0, tf, y0, h):
    t_values = np.arange(t0, tf, h)
    y_values = [y0]

    for t in t_values[:-1]:
        y_n = y_values[-1]
        k1 = np.array(f(t, y_n))
        k2 = np.array(f(t + h/2, y_n + 0.5*h*k1))
        k3 = np.array(f(t + h, y_n - k1 + 2*k2))
        y_n1 = y_n + (1/6) * h * (k1 + 4*k2 + k3)
        y_values.append(y_n1)

    return t_values, np.array(y_values).T

# Função que implementa o método de Runge-Kutta de ordem 4
def runge_kutta_ordem_4(f, t0, tf, y0, h):
    t_values = np.arange(t0, tf, h)
    y_values = [y0]

    for t in t_values[:-1]:
        y_n = y_values[-1]
        k1 = np.array(f(t, y_n))
        k2 = np.array(f(t + h/2, y_n + 0.5*h*k1))
        k3 = np.array(f(t + h/2, y_n + 0.5*h*k2))
        k4 = np.array(f(t + h, y_n + h*k3))
        y_n1 = y_n + (1/6) * h * (k1 + 2*k2 + 2*k3 + k4)
        y_values.append(y_n1)

    return t_values, np.array(y_values).T
```

Cada método utiliza uma abordagem ligeiramente diferente para calcular as soluções numéricas das EDOs, com diferentes níveis de precisão. A escolha do método depende das necessidades específicas do problema a ser resolvido e dos requisitos de precisão. Em geral, quanto maior a ordem do método, maior a precisão, mas também pode ser computacionalmente mais custoso.

### **SIR com métodos númericos**

```python
import numpy as np
import matplotlib.pyplot as plt

# Função que define o sistema de equações diferenciais do modelo SIR
def sir_model(t, y, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I
    dRdt = gamma * I
    return [dSdt, dIdt, dRdt]

# Parâmetros do modelo SIR
beta = 0.5  # Taxa de transmissão
gamma = 0.1  # Taxa de recuperação

# Condições iniciais
S0 = 0.99  # População suscetível inicial
I0 = 0.01  # População infectada inicial
R0 = 0.0  # População recuperada inicial

# População recuperada inicial
y0 = [S0, I0, R0]

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 100.0  # Tempo final
h = 0.1  # Tamanho do passo

# Resolvendo o sistema de equações diferenciais usando o método escolhido
method = euler_melhorado  # Método escolhido (euler_melhorado, runge_kutta_ordem_3, runge_kutta_ordem_4)
t_values, y_values = method(lambda t, y: sir_model(t, y, beta, gamma), t0, tf, y0, h)

# Plotando o gráfico da solução
plt.plot(t_values, y_values[0], label='Susceptíveis')
plt.plot(t_values, y_values[1], label='Infectados')
plt.plot(t_values, y_values[2], label='Recuperados')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SIR')
plt.grid(True)
plt.legend()
plt.show()
```

A principal diferença em relação ao primeiro código é que este código implementa a resolução das EDOs usando métodos numéricos, enquanto o primeiro código utilizava a função solve_ivp do SciPy para resolver as EDOs. Aqui, temos a opção de escolher o método numérico a ser usado para a resolução das EDOs (como o Método de Euler Melhorado, Runge-Kutta de Ordem 3 e Runge-Kutta de Ordem 4), enquanto o primeiro código usava um único método. Na execução acima, utilizamos o método do Euler melhorado.

### **SIRD**

Este código implementa um modelo SIRD (Susceptíveis-Infectados-Recuperados-Falecidos) para simular a propagação de uma doença infecciosa ao longo do tempo, utilizando uma implementação numérica dos métodos de resolução de equações diferenciais ordinárias (EDOs), como o Método de Euler Melhorado, Runge-Kutta de Ordem 3 e Runge-Kutta de Ordem 4.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função que define o sistema de equações diferenciais do modelo SIRD
def sird_model(t, y, beta, gamma, delta):
    S, I, R, D = y
    dSdt = -beta * S * I
    dIdt = beta * S * I - gamma * I - delta * I
    dRdt = gamma * I
    dDdt = delta * I
    return [dSdt, dIdt, dRdt, dDdt]

# Parâmetros do modelo SIRD
beta = 0.5  # Taxa de transmissão
gamma = 0.1  # Taxa de recuperação
delta = 0.05  # Taxa de mortalidade

# Condições iniciais
S0 = 0.99  # População suscetível inicial
I0 = 0.01  # População infectada inicial
R0 = 0.0  # População recuperada inicial
D0 = 0.0  # População falecida inicial
y0 = [S0, I0, R0, D0]

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 100.0  # Tempo final
h = 0.1  # Tamanho do passo

# Resolvendo o sistema de equações diferenciais usando o método escolhido
method = runge_kutta_ordem_3  # Método escolhido (euler_melhorado, runge_kutta_ordem_3, runge_kutta_ordem_4)
t_values, y_values = method(lambda t, y: sird_model(t, y, beta, gamma, delta), t0, tf, y0, h)

# Plotando o gráfico da solução
plt.plot(t_values, y_values[0], label='Suscetíveis')
plt.plot(t_values, y_values[1], label='Infectados')
plt.plot(t_values, y_values[2], label='Recuperados')
plt.plot(t_values, y_values[3], label='Mortos')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SIRD')
plt.grid(True)
plt.legend()
plt.show()
```

A principal diferença em relação ao primeiro código é a incorporação do modelo SIRD, que inclui uma população de Falecidos (D) e a taxa de mortalidade (delta). Isso permite simular a propagação da doença e a mortalidade associada. Além disso, assim como anteriormente, temos a opção de escolher o método numérico a ser usado para a resolução das EDOs, e aqui utilizamos o Runge-Kutta de ordem 3.

Sobre o gráfico, vemos que a curva que representa a população suscetível (S) mostra uma diminuição gradual à medida que os indivíduos se tornam infectados.

A curva que representa a população infectada (I) inicialmente aumenta devido à transmissão da doença (termo positivo de betaSI) e diminui posteriormente devido à recuperação (termo negativo de gammaI) e à mortalidade (termo negativo de deltaI).

A curva que representa a população recuperada (R) mostra um aumento gradual à medida que os indivíduos se recuperam da doença. Isso ocorre devido à taxa de recuperação (gamma).

A curva que representa a população falecida (D) aumenta à medida que os indivíduos infectados falecem devido à taxa de mortalidade (delta), sendo que o número de óbitos é diretamente proporcional à quantidade de infectados.

```python
# Novos parâmetros do modelo SIRD
beta = 0.6  # Taxa de transmissão
gamma = 0.2  # Taxa de recuperação
delta = 0.05  # Taxa de mortalidade

# Novas condições iniciais
S0 = 0.9  # População suscetível inicial
I0 = 0.1  # População infectada inicial
R0 = 0.0  # População recuperada inicial
D0 = 0.0  # População falecida inicial

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 200.0  # Tempo final
h = 0.1  # Tamanho do passo

# Resolvendo o sistema de equações diferenciais usando o método escolhido
method = runge_kutta_ordem_3  # Método escolhido (euler_melhorado, runge_kutta_ordem_3, runge_kutta_ordem_4)
t_values, y_values = method(lambda t, y: sird_model(t, y, beta, gamma, delta), t0, tf, [S0, I0, R0, D0], h)

# Plotando o gráfico da solução
plt.plot(t_values, y_values[0], label='Suscetíveis')
plt.plot(t_values, y_values[1], label='Infectados')
plt.plot(t_values, y_values[2], label='Recuperados')
plt.plot(t_values, y_values[3], label='Mortos')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SIRD com Novos Parâmetros')
plt.grid(True)
plt.legend()
plt.show()
```

A principal diferença em relação ao gráfico anterior é a combinação de parâmetros que resulta em uma taxa de transmissão mais alta, uma taxa de recuperação mais alta e uma taxa de mortalidade. Isso leva a uma propagação mais rápida da doença, um pico mais pronunciado de infectados e uma diminuição mais rápida na população suscetível.
Além disso, a proporção inicial de infectados (I0 = 0.1) influencia a dinâmica do modelo, levando a um aumento mais rápido dos casos iniciais.


Em resumo, o novo conjunto de parâmetros resulta em uma dinâmica do modelo SIRD que pode representar uma situação em que a doença se propaga mais rapidamente, atingindo um pico mais alto e diminuindo mais rapidamente, devido à combinação de uma alta taxa de transmissão, uma alta taxa de recuperação e uma taxa de mortalidade.

### **SEIR**

Aqui, implementamos o modelo SEIR (Susceptíveis-Expostos-Infectados-Recuperados) para simular a propagação de uma doença infecciosa ao longo do tempo e escolhemos o método de Runge-Kutta de ordem 4.

Neste modelo SEIR, há uma nova categoria chamada "Expostos (E)", que representa indivíduos que foram expostos à doença, mas ainda não são infecciosos. Isso reflete um período de incubação em que as pessoas estão em risco, mas ainda não estão transmitindo a doença.

```python
import numpy as np
import matplotlib.pyplot as plt

# Função que define o sistema de equações diferenciais do modelo SEIR
def seir_model(t, y, beta, sigma, gamma):
    S, E, I, R = y
    dSdt = -beta * S * I
    dEdt = beta * S * I - sigma * E
    dIdt = sigma * E - gamma * I
    dRdt = gamma * I
    return [dSdt, dEdt, dIdt, dRdt]

# Parâmetros do modelo SEIR
beta = 0.5  # Taxa de transmissão
sigma = 0.1  # Taxa de incubação
gamma = 0.05  # Taxa de recuperação

# Condições iniciais
S0 = 0.99  # População suscetível inicial
E0 = 0.01  # População exposta inicial
I0 = 0.0  # População infectada inicial
R0 = 0.0  # População recuperada inicial
y0 = [S0, E0, I0, R0]

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 100.0  # Tempo final
h = 0.1  # Tamanho do passo

# Resolvendo o sistema de equações diferenciais usando o método escolhido
method = runge_kutta_ordem_4  # Método escolhido (euler_melhorado, runge_kutta_ordem_3, runge_kutta_ordem_4)
t_values, y_values = method(lambda t, y: seir_model(t, y, beta, sigma, gamma), t0, tf, y0, h)

# Plotando o gráfico da solução
plt.plot(t_values, y_values[0], label='Suscetíveis')
plt.plot(t_values, y_values[1], label='Expostos')
plt.plot(t_values, y_values[2], label='Infectados')
plt.plot(t_values, y_values[3], label='Recuperados')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SEIR')
plt.grid(True)
plt.legend()
plt.show()
```

Essencialmente, o modelo SEIR permite modelar o período de incubação e o tempo entre a exposição e a infecção, o que o torna mais adequado para representar a dinâmica de doenças infecciosas em que há um período latente antes que os indivíduos se tornem infecciosos.

A curva de "Expostos" (E) mostra um aumento inicial, refletindo o período de incubação, antes de diminuir. A curva de "Infectados" (I) aumenta após o período de incubação e diminui quando as pessoas se recuperam. A curva de "Recuperados" (R) aumenta à medida que os infectados se recuperam.





```python
# Novos parâmetros do modelo SEIR
beta = 0.4  # Taxa de transmissão
sigma = 0.2  # Taxa de incubação
gamma = 0.1  # Taxa de recuperação

# Novas condições iniciais
S0 = 0.95  # População suscetível inicial
E0 = 0.02  # População exposta inicial
I0 = 0.03  # População infectada inicial
R0 = 0.0  # População recuperada inicial

# Intervalo de tempo
t0 = 0.0  # Tempo inicial
tf = 150.0  # Tempo final
h = 0.1  # Tamanho do passo

# Resolvendo o sistema de equações diferenciais usando o método escolhido
method = runge_kutta_ordem_4  # Método escolhido (euler_melhorado, runge_kutta_ordem_3, runge_kutta_ordem_4)
t_values, y_values = method(lambda t, y: seir_model(t, y, beta, sigma, gamma), t0, tf, [S0, E0, I0, R0], h)

# Plotando o gráfico da solução
plt.plot(t_values, y_values[0], label='Suscetíveis')
plt.plot(t_values, y_values[1], label='Expostos')
plt.plot(t_values, y_values[2], label='Infectados')
plt.plot(t_values, y_values[3], label='Recuperados')
plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Modelo SEIR com Novos Parâmetros')
plt.grid(True)
plt.legend()
plt.show()
```

A principal diferença em relação ao código anterior (modelo SEIR com diferentes parâmetros) é a taxa de transmissão mais baixa (beta = 0.4), que resulta em uma propagação mais lenta da doença.

O período de incubação é mais longo devido à taxa de incubação (sigma = 0.2), o que é evidenciado pela curva de expostos.

A proporção de suscetíveis no início da simulação é maior, resultando em menos indivíduos infectados no início.

Em resumo, o novo conjunto de parâmetros resulta em uma dinâmica do modelo SEIR que representa uma propagação mais lenta da doença e um período de incubação mais longo. A taxa de transmissão reduzida e a maior proporção de suscetíveis resultam em uma menor taxa de infecção no início da simulação.

# **Conclusão**

#### Os modelos epidemiológicos SIR, SIRD, SEIR e SEIRD são ferramentas essenciais para entender e prever a propagação de doenças infecciosas em uma população. Cada modelo oferece uma abordagem única para representar as dinâmicas da doença e levar em consideração diferentes estágios da infecção. Resumidamente, temos:

<br>

#### **Modelo SIR (Susceptíveis-Infectados-Recuperados):**
- #### Divide a população em três compartimentos: suscetíveis, infectados e recuperados.
- #### Pressupõe que uma vez recuperado, um indivíduo adquire imunidade e não pode ser reinfectado.

#### **Modelo SIRD (Susceptíveis-Infectados-Recuperados-Falecidos):**
- #### Expande o modelo SIR ao incluir um compartimento para os óbitos.
- #### Considera a mortalidade devido à doença infecciosa.

#### **Modelo SEIR (Susceptíveis-Expostos-Infectados-Recuperados):**
- #### Introduz o compartimento de "Expostos", representando indivíduos expostos à doença, mas não infecciosos.
- #### Modela o período de incubação, durante o qual os indivíduos não são contagiosos.

#### **Modelo SEIRD (Susceptíveis-Expostos-Infectados-Recuperados-Falecidos):**
- #### Combinando os elementos dos modelos SEIR e SIRD, inclui a mortalidade devido à doença e o período de exposição.

<br>

#### Em relação aos métodos numéricos de Euler e Runge-Kutta, eles desempenham um papel fundamental na solução desses modelos de equações diferenciais. Vamos destacar as principais características desses métodos:

<br>

#### **Método de Euler:**
- #### É um método de passo único e simples de implementar.
- #### Divide o intervalo de tempo em pequenos passos e estima a próxima solução com base na derivada local.
- #### Pode ser menos preciso, especialmente em situações com alta variação ou respostas não lineares.

#### **Método de Runge-Kutta:**
- #### Inclui diferentes variações, como ordem 3 (Runge-Kutta de ordem 3) e ordem 4 (Runge-Kutta de ordem 4), que oferecem maior precisão.
- #### Usa uma abordagem de múltiplos estágios para calcular a próxima solução.
- #### Geralmente é mais preciso do que o método de Euler e pode ser adaptado para uma ampla gama de problemas.

<br>

#### Em resumo, os modelos epidemiológicos e os métodos numéricos desempenham um papel crítico na previsão e compreensão da propagação de doenças infecciosas. A escolha do modelo e do método numérico depende das características específicas da doença, da disponibilidade de dados e dos recursos computacionais. A combinação dessas ferramentas permite aos epidemiologistas e pesquisadores melhorar a tomada de decisões, avaliar estratégias de intervenção e entender o impacto de diferentes cenários epidemiológicos. Essas abordagens desempenham um papel vital em situações de pandemias e na saúde pública global.
