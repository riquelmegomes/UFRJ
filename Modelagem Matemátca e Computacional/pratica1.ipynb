# **Sistema de Lorenz**

> #### O Sistema de Lorenz, nomeado após o meteorologista Edward Lorenz, é um conjunto de equações diferenciais não lineares que descrevem o comportamento caótico de um sistema dinâmico. Esse sistema é amplamente reconhecido por ser um dos primeiros exemplos de um sistema determinístico que exibe sensibilidade extrema às condições iniciais, o que leva ao comportamento caótico.

\
#### As equações de Lorenz foram originalmente formuladas para modelar a convecção atmosférica simplificada, mas ao longo do tempo, elas foram adotadas para descrever uma variedade de fenômenos, desde sistemas climáticos até oscilações químicas e físicas. O sistema consiste em três equações diferenciais acopladas, que representam as variações temporais de três variáveis dependentes: $x$, $y$ e $z$.


#### As equações de Lorenz são as seguintes:
1. ### $dx/dt = σ(y - x)$
2. ### $dy/dt = x(ρ - z) - y$
3. ### $dz/dt = xy - βz$

#### Onde $x$, $y$ e $z$ são as variáveis do sistema. $σ$, $ρ$ e $β$ são os parâmetros do sistema, que determinam o comportamento do sistema e a sua sensibilidade às condições iniciais e t é o tempo.
<br>

#### Uma das características mais marcantes do sistema de Lorenz é o fenômeno conhecido como "efeito borboleta". Isso se refere à sensibilidade extrema às condições iniciais, em que uma pequena alteração nos valores iniciais das variáveis pode resultar em comportamentos completamente diferentes no longo prazo. Devido à natureza caótica do sistema de Lorenz, ele não é previsível no longo prazo, mesmo que as equações que o descrevem sejam determinísticas e completamente conhecidas. Isso significa que, para sistemas caóticos como esse, pequenas incertezas nas medições iniciais ou nos parâmetros podem resultar em predições substancialmente diferentes ao longo do tempo.
<br>

#### Em resumo, o sistema de Lorenz é um exemplo clássico de sistema dinâmico caótico que ajudou a estabelecer as bases da teoria do caos. Suas equações ilustram como sistemas aparentemente simples podem exibir comportamento complexo e imprevisível devido à sua sensibilidade extrema às condições iniciais.

### **Implementação em Python para o sistema de Lorenz**
(código disponibilizado via Classroom)

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Aug 13 21:51:58 2023

@author: amaury
"""

import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Definição das equações diferenciais do sistema de Lorenz
def lorenz(t, xyz, sigma, rho, beta):
    x, y, z = xyz
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Condições iniciais
initial_conditions = [0, 1, 1.05]

# Intervalo de tempo para a integração
t_span = (0, 25)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Parâmetros do sistema de Lorenz
sigma = 10
rho = 28
beta = 8/3

# Resolvendo o sistema de Lorenz
solution = solve_ivp(lorenz, t_span, initial_conditions, args=(sigma, rho, beta), t_eval=t_eval)

# Plotando o gráfico da solução
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2], lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title(f'Solução do Sistema de Lorenz\nCondições iniciais: {initial_conditions}')
plt.show()
```

#### É possível observar que o gráfico, construído como resultado do código do sistema de Lorenz com condições iniciais $[0, 1, 1.05]$ e parâmetros $σ = 10$, $ρ = 28$ e $β = 8/3$, representa visualmente uma curva que aparenta circundar ao redor de dois pontos centrais, esses são os **atratores**.

## **Atratores**

> #### Atratores são conceitos fundamentais na teoria dos sistemas dinâmicos e no estudo do caos. Eles são padrões ou regiões no espaço de fase de um sistema dinâmico para os quais as trajetórias das variáveis do sistema tendem a convergir ou permanecer ao longo do tempo. Os atratores são representações gráficas ou geométricas de comportamentos dinâmicos complexos que podem ser úteis para visualizar e entender a evolução temporal de sistemas dinâmicos. O conceito de atratores é essencial na teoria do caos, uma vez que muitos sistemas caóticos exibem atratores caóticos que revelam a natureza complexa e imprevisível do comportamento desses sistemas.

\
#### Existem três tipos principais de atratores: atratores fixos (ou pontos fixos), atratores periódicos e atratores caóticos.

1. #### **Atratores Fixos (Pontos Fixos)**: São pontos ou regiões no espaço de fase para os quais as trajetórias do sistema convergem e permanecem estáveis. Isso significa que, independentemente das condições iniciais, o sistema acabará alcançando esse estado. Esses atratores representam estados de equilíbrio ou estabilidade para o sistema.

2. #### **Atratores Periódicos:** São regiões no espaço de fase nas quais as trajetórias do sistema repetem ciclicamente padrões ao longo do tempo. Em vez de convergir para um ponto fixo, o sistema oscila em torno de uma trajetória fechada. A frequência e a forma da oscilação dependem das características do atrator periódico.

3. #### **Atratores Caóticos:** São regiões complexas e não lineares no espaço de fase onde as trajetórias do sistema exibem comportamento caótico. As trajetórias em um atrator caótico são altamente sensíveis às condições iniciais, o que significa que pequenas variações nas condições iniciais levam a evoluções divergentes das trajetórias. A geometria fractal é frequentemente associada aos atratores caóticos, tornando-os altamente complexos e aparentemente irregulares.
<br>

## **Atratores de Lorenz**

> #### Os atratores de Lorenz têm uma forma distintiva de "borboleta", que é o motivo pelo qual eles são frequentemente chamados de "Atratores de Borboleta". Essa forma peculiar é um resultado da **sensibilidade extrema** às condições iniciais, o que significa que mesmo pequenas variações nas condições iniciais do sistema podem levar a trajetórias completamente diferentes ao longo do tempo. Em termos simples, isso sugere que mesmo a menor alteração nas medições iniciais pode resultar em previsões drasticamente divergentes à medida que o tempo avança.



## **Sensibilidade do sistema de Lorenz**

#### Manifestada como uma característica fundamental que distingue esse sistema dinâmico e contribui para sua natureza caótica e imprevisível, a sensibilidade do sistema de Lorenz se manifesta de forma extrema às condições iniciais. Isso significa que mesmo as mínimas variações nas medições iniciais das variáveis do sistema podem levar a resultados radicalmente diferentes à medida que o tempo avança. A imagem icônica do "efeito borboleta", frequentemente associada ao sistema de Lorenz, ilustra essa ideia de maneira poética: o bater de asas de uma borboleta em um local distante pode, teoricamente, desencadear uma série de eventos que, ao longo do tempo, resultariam em mudanças significativas no sistema.
<br>

#### Essa sensibilidade às condições iniciais é uma característica central do caos determinístico. Apesar de o sistema de Lorenz ser governado por equações precisas, a imprevisibilidade decorrente das pequenas variações nas condições iniciais desafia a ideia tradicional de previsibilidade. Mesmo que um observador conheça todas as equações que governam o sistema com precisão, a capacidade de fazer previsões precisas a longo prazo é severamente limitada pela falta de conhecimento das condições iniciais exatas.

# **Tarefa 2**

#### Para essa tarefa, criei a função em phyton "derivada_disc" que recebe uma lista de valores e calcula a derivada discreta desses usando a fórmula de **diferenças finitas** e uma variação horizontal 'deltax' dada como entrada da função. Em seguida, ela constrói e mostra os gráficos com a lista de valores iniciais e com as derivadas calculadas. A partir disso, utilizando a função "linregress" da biblioteca SciPy para efetuar **regressão linear** com o **método dos mínimos quadrados**, obtemos, com o uso de um modelo linear, uma lei aproximada que representa a lista de resultados que calculamos. Os conceitos e métodos matemáticos citados e utilizados no código são brevemente explicados mais adiante.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

def derivada_disc(lista, deltax):
  tam = len(lista)
  resultado = []

  # Cria lista de valores em x com 'deltax' de variação entre eles
  x = np.arange(0, tam*deltax, deltax)

  # Plota o gráfico dos pontos recebidos
  plt.figure()
  plt.scatter(x, lista, label = "Valores da lista de dados")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.show()
  print()

  # Calcula a derivada discreta usando a lista de dados
  for i in range(0, tam-1):
    resultado.append(lista[i+1] - lista[i] / deltax)
  i += 1
  resultado.append(lista[i] - lista[i-1] / deltax) # a derivada do último valor é aproximada de forma regressiva

  # Obtém o modelo que aproxima a derivada, usando regressão linear e mínimos quadrados
  a, b, r2, p_value, std_err = linregress(x, resultado)

  # Plota o gráfico do resultado
  plt.figure()
  plt.plot(x, a*x+b, label = "Reta ajustada com linregress")
  plt.scatter(x, resultado, label = "Derivada discreta dos dados")
  plt.xlabel("x")
  plt.ylabel("derivada discreta")
  plt.legend()
  plt.show()

  # Printa os valores de a e b obtidos
  print(f"\nModelo da regressão linear: y = ax + b\n a = {a}, b = {b}")
```

#### Como exemplo, utilizei um dataset real, retirado do site Kaggle, que apresenta um registro abrangente da renda média por pessoa no Canadá ao longo de vários anos. Mais especificamente, utilizei a coluna dos valores de renda disponíveis nesse conjunto de dados como um caso de lista de entrada para minha função citada, de forma que é possível utilizar o método para testar diversas outras listas. Como resultado, encontrei valores aproximados para $a$ e $b$ do modelo linear $y = ax + b$.

```python
# importação dos dados de exemplo
lista = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Canada-per-capita-income.csv")["income"]

# Função que obtém a lei que representa a lista
derivada_disc(lista, deltax=2)
```

## **Métodos matemáticos utilizados:**

### **Método das diferenças finitas**

> #### O método das diferenças finitas é uma abordagem numérica usada para resolver equações diferenciais em uma variedade de campos. Ele envolve dividir um domínio contínuo em pontos discretos e substituir derivadas por diferenças entre valores em pontos vizinhos. Isso transforma as equações diferenciais em equações algébricas que podem ser resolvidas numericamente. O processo inclui discretizar o domínio, aproximar derivadas, substituir essas aproximações na equação original e resolver o sistema de equações resultante. Esse método é valioso quando não há soluções analíticas disponíveis ou quando a equação não pode ser resolvida por outros métodos numéricos, sendo frequentemente aplicado em simulações computacionais e análises de problemas complexos. A precisão das soluções depende da densidade da grade e da escolha das diferenças utilizadas, exigindo considerações sobre estabilidade numérica para obter resultados confiáveis.

<br>

### **Regressão Linear**
> #### A regressão linear é um método estatístico usado para entender e modelar a relação entre duas variáveis. Ele busca encontrar uma linha reta que melhor se ajusta aos dados, minimizando a distância entre os pontos e a linha. Essa linha é chamada de "linha de regressão" e é usada para fazer previsões ou entender a tendência dos dados. A equação da regressão linear é da forma $y = mx + b$, onde "$y$" é a variável dependente, "$x$" é a variável independente, "$m$" é a inclinação da linha e "$b$" é a intercepção no eixo $y$. A regressão linear é amplamente usada em análise estatística, pesquisa de mercado, economia e ciências naturais para entender e prever relações entre variáveis.

<br>

### **Método dos mínimos quadrados**
> #### Mínimos quadrados é um método matemático utilizado para encontrar a melhor aproximação de uma linha ou curva aos dados observados. O objetivo é minimizar a soma dos quadrados das diferenças entre os valores reais e os valores estimados pela linha ou curva. Isso envolve ajustar os parâmetros da linha ou curva de tal forma que a discrepância entre os dados e as previsões seja reduzida ao mínimo. O método é frequentemente usado em regressão, onde a linha de melhor ajuste é encontrada para representar uma relação entre variáveis. O princípio dos mínimos quadrados é amplamente aplicado em diversas áreas, como análise de dados, estatísticas, engenharia e economia, para encontrar modelos que melhor descrevam as relações subjacentes nos dados observados.
