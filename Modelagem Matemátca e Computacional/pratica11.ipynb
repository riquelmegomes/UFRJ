# **Teorema Central do Limite**

#### O Teorema Central do Limite (TCL) é um dos princípios fundamentais da estatística e da teoria das probabilidades. Ele descreve o comportamento das médias de uma grande quantidade de variáveis aleatórias independentes e identicamente distribuídas (IID) à medida que o tamanho da amostra aumenta. O TCL é crucial porque permite entender como a distribuição de médias amostrais se comporta, independentemente da distribuição das variáveis originais.

<br>

#### A afirmação principal do Teorema Central do Limite é a seguinte:

> #### Dada uma amostra suficientemente grande de variáveis aleatórias independentes e identicamente distribuídas (IID), a distribuição das médias amostrais se aproxima de uma **distribuição normal** (gaussiana), independentemente da forma da distribuição original das variáveis.

<br>

#### Isso significa que, mesmo que as variáveis originais não sigam uma distribuição normal, a média das amostras tende a se aproximar de uma distribuição normal à medida que o tamanho da amostra aumenta. A distribuição normal é caracterizada pela sua forma de sino, e muitos resultados estatísticos e inferências são baseados nessa distribuição, o que torna o TCL um resultado crucial na estatística.

<br>

#### O TCL tem várias implicações práticas:

1. #### Permite a aplicação de métodos de inferência estatística, como testes de hipóteses e intervalos de confiança, mesmo quando a distribuição das variáveis originais é desconhecida.

2. #### É a base para a construção de modelos probabilísticos e previsões em muitas áreas da ciência, negócios e engenharia.

3. #### Ajuda a explicar por que a distribuição normal é tão prevalente na natureza, uma vez que muitos fenômenos complexos podem ser modelados por meio da média de várias variáveis aleatórias independentes.

4. #### O TCL é frequentemente usado em aplicações práticas, como na análise de dados, em que amostras de dados são frequentemente usadas para inferir informações sobre populações maiores.

<br>

#### Em resumo, o Teorema Central do Limite descreve como a média de um grande número de observações de variáveis aleatórias tende a se comportar, o que é uma ferramenta fundamental na estatística e na análise de dados.

# **Lei dos Grandes Números**

#### A Lei dos Grandes Números (LLN, da sigla em inglês "Law of Large Numbers") é um princípio fundamental na teoria das probabilidades e na estatística. Ela descreve o comportamento de médias amostrais à medida que o tamanho da amostra aumenta. A LLN é fundamentalmente importante para entender como a estatística funciona em situações do mundo real

<br>

#### A Lei dos Grandes Números afirma basicamente o seguinte:

> #### À medida que o tamanho da amostra aumenta, a média das observações se aproxima cada vez mais da média da população da qual as observações foram tiradas.

<br>

#### Isso significa que, à medida que você coleta mais dados, a média desses dados se torna uma estimativa cada vez melhor da média real da população. Em outras palavras, à medida que você aumenta o tamanho da amostra, a média amostral tende a convergir para a média populacional.

<br>

#### A Lei dos Grandes Números pode ser formulada em duas versões principais:

1. #### **LLN Fraca:** Esta versão afirma que, à medida que o tamanho da amostra aumenta, a média das observações converge em probabilidade para a média populacional. Em termos simples, a probabilidade de a média amostral se afastar da média populacional diminui à medida que o tamanho da amostra aumenta.

2. #### **LLN Forte:** Esta versão é mais forte e afirma que, à medida que o tamanho da amostra aumenta, a média das observações converge quase certamente para a média populacional. Isso significa que, com probabilidade igual a $1$, a média amostral se aproxima da média populacional à medida que o tamanho da amostra aumenta.

<br>

#### A Lei dos Grandes Números é fundamental para a estatística e tem várias implicações práticas:

- #### Ela fornece uma base para a confiabilidade da média amostral como estimativa da média populacional.
- #### Ajuda a explicar por que é possível fazer previsões com base em dados amostrais.
- #### É essencial para a análise estatística e para a validade de muitos métodos estatísticos e inferências.
- #### É amplamente utilizada em finanças, engenharia, ciência, e em muitas outras áreas para a tomada de decisões baseadas em dados.

<br>

#### Em resumo, a Lei dos Grandes Números é um conceito fundamental que descreve como médias amostrais se aproximam da média populacional à medida que o tamanho da amostra aumenta, tornando-a essencial para a análise estatística e a tomada de decisões baseadas em dados.

# **Distribuição de Pareto**

#### A Distribuição de Pareto, também conhecida como a Lei de Pareto, é um modelo estatístico que descreve a distribuição de uma grande variedade de fenômenos econômicos, sociais e naturais, onde a maioria dos resultados ou efeitos é concentrada em uma minoria de causas ou elementos. Ela é nomeada em homenagem a Vilfredo Pareto, um economista italiano que observou pela primeira vez esse padrão de desigualdade em sua análise da distribuição de renda na Itália no final do século XIX.

<br>

#### A Distribuição de Pareto é caracterizada por sua propriedade de cauda longa, o que significa que uma grande proporção dos resultados está concentrada nas caudas da distribuição, com uma minoria de resultados representando a maioria dos efeitos. Em termos simples, a regra 80/20 frequentemente se aplica à Distribuição de Pareto, onde aproximadamente 80% dos efeitos resultam de cerca de 20% das causas ou elementos.

<br>

#### As principais características da Distribuição de Pareto incluem:

1. #### **Cauda longa:** A cauda da distribuição se estende indefinidamente para a direita, o que significa que pode haver valores extremamente altos.

2. #### **Parâmetro de forma:** A Distribuição de Pareto é caracterizada por um parâmetro de forma (também chamado de índice de Pareto) que determina a inclinação da cauda. Quanto menor o valor desse parâmetro, mais acentuada é a cauda longa.

3. #### **Aplicação em diversos campos:** A Distribuição de Pareto é observada em muitos contextos, como distribuição de renda, distribuição de riqueza, sucesso em negócios, desempenho de produtos, classificação de clientes, frequência de palavras em textos, falhas de sistemas, entre outros.

4. #### **Princípio 80/20:** A regra 80/20 é frequentemente associada à Distribuição de Pareto, onde 20% das causas ou elementos são responsáveis por 80% dos resultados. No entanto, os números exatos podem variar.

<br>

#### A Distribuição de Pareto é frequentemente usada em análises de gestão, tomada de decisões e planejamento estratégico. Ao identificar e priorizar os elementos ou causas que têm o maior impacto em um sistema ou processo, as organizações podem direcionar seus recursos de maneira mais eficaz.

<br>

#### É importante ressaltar que a Distribuição de Pareto não é a única distribuição estatística que exibe caudas longas, mas é uma das mais conhecidas e amplamente aplicadas em diferentes áreas.

### **Teorema Central do Limite**

Este código em Python ilustra o Teorema Central do Limite (TCL) gerando amostras de uma distribuição específica, calculando as médias dessas amostras e exibindo um histograma das médias amostrais. Neste caso, estamos usando uma distribuição normal com média (mu) zero, desvio padrão (sigma) um e definimos o número de amostras a serem geradas (num_samples) e o tamanho de cada amostra (sample_size).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da distribuição original
mu = 0  # média
sigma = 1  # desvio padrão

# Número de amostras e tamanho de cada amostra
num_samples = 1000
sample_size = 30

# Gere amostras de uma distribuição qualquer (normal neste exemplo)
samples = np.random.normal(mu, sigma, size=(num_samples, sample_size))

# Calcule as médias de cada amostra
sample_means = np.mean(samples, axis=1)

# Crie um histograma das médias amostrais
plt.hist(sample_means, bins=50, density=True, alpha=0.6, color='g', edgecolor='black')

# Adicione uma linha vertical na média da população original
plt.axvline(mu, color='red', linestyle='dashed', linewidth=2, label='Média da População')

# Adicione legendas e título
plt.title('Teorema Central do Limite')
plt.xlabel('Média Amostral')
plt.ylabel('Densidade de Probabilidade')
plt.legend()

# Exiba o gráfico
plt.show()
```

O resultado gráfico é um histograma das médias amostrais, onde é possível observar visualmente como a distribuição das médias amostrais se aproxima de uma distribuição normal à medida que o número de amostras aumenta, corroborando o Teorema Central do Limite. A linha vermelha pontilhada representa a média da população original, destacando como as médias amostrais se concentram em torno dessa média à medida que mais amostras são consideradas.

### **Lei dos Grandes Números**

Este código gera amostras de uma distribuição normal, calcula as médias amostrais para diferentes tamanhos de amostra e exibe um gráfico que demonstra como as médias amostrais convergem para a média da população à medida que o tamanho da amostra aumenta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da distribuição original
mu = 3  # média
sigma = 2  # desvio padrão

# Número de amostras e tamanhos de amostra
num_samples = 1000
max_sample_size = 1000

# Gere amostras de uma distribuição qualquer (normal neste exemplo)
population = np.random.normal(mu, sigma, size=num_samples)

# Inicialize arrays para armazenar as médias amostrais
sample_means = np.zeros(max_sample_size)

# Gere médias amostrais para diferentes tamanhos de amostra
for sample_size in range(1, max_sample_size + 1):
    samples = np.random.choice(population, size=sample_size)
    sample_means[sample_size - 1] = np.mean(samples)

# Crie um gráfico das médias amostrais em função do tamanho da amostra
plt.plot(range(1, max_sample_size + 1), sample_means, label='Médias Amostrais')
plt.axhline(mu, color='red', linestyle='dashed', linewidth=2, label='Média da População')

# Adicione legendas e título
plt.title('Lei dos Grandes Números')
plt.xlabel('Tamanho da Amostra')
plt.ylabel('Média Amostral')
plt.legend()

# Exiba o gráfico
plt.show()
```

Os resultados gráficos mostrarão uma curva que representa como a média amostral varia à medida que o tamanho da amostra aumenta. A linha vermelha pontilhada representa a média da população original (3 no caso). Conforme o tamanho da amostra aumenta, a média amostral converge para a média da população, demonstrando graficamente a Lei dos Grandes Números. Isso ilustra que, em média, a estimativa da média da população se torna mais precisa à medida que mais dados são considerados.

### **Distribuição de Pareto**

Este código gera um gráfico que ilustra a Distribuição de Pareto com base nos parâmetros especificados. Estes são os parâmetros da Distribuição de Pareto: o parâmetro de forma (alpha) determina a inclinação da cauda da distribuição, enquanto o valor mínimo (xmin) é o ponto a partir do qual os dados começam a ser gerados.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da Distribuição de Pareto
alpha = 2.5  # Parâmetro de forma (shape parameter)
xmin = 1  # Valor mínimo

# Gere dados seguindo a Distribuição de Pareto
data = np.random.pareto(alpha, 1000) + xmin

# Crie um histograma dos dados
plt.hist(data, bins=30, density=True, alpha=0.7, color='b', edgecolor='black')

# Adicione uma linha vertical no valor mínimo
plt.axvline(xmin, color='red', linestyle='dashed', linewidth=2, label='Valor Mínimo')

# Adicione legendas e título
plt.title('Distribuição de Pareto')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()

# Exiba o gráfico
plt.show()
```

Como resultado, podemos ver um gráfico que mostra a forma característica da Distribuição de Pareto, com a maioria dos valores concentrados à esquerda e uma cauda longa se estendendo para a direita. A linha vermelha pontilhada destaca o valor mínimo (xmin). Esse tipo de distribuição é frequentemente observado em fenômenos em que a maioria dos resultados está concentrada em um número relativamente pequeno de eventos.

O código abaixo demonstra diferentes métodos para gerar números aleatórios e UUIDs (identificadores únicos universais).

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Oct 23 07:22:27 2023

@author: amaury
"""
import random
import numpy as np
import secrets
import uuid

# Método 1: Usando o módulo random
print("Usando o módulo random")
random_number = random.random()
print(f"Número aleatório (random.random()): {random_number}")

# Método 2: Usando o módulo random para gerar números inteiros
random_int = random.randint(1, 100)
print(f"Número aleatório inteiro (random.randint()): {random_int}")

# Método 3: Usando o módulo random com seed
random.seed(42)
random_with_seed = random.random()
print(f"Número aleatório com seed (random.seed()): {random_with_seed}")

# Método 4: Usando o módulo numpy
print("\nUsando o módulo numpy")
numpy_random = np.random.rand()
print(f"Número aleatório (numpy.random.rand()): {numpy_random}")

# Método 5: Usando o módulo secrets para números criptograficamente seguros
print("\nUsando o módulo secrets")
crypto_random = secrets.randbelow(100)
print(f"Número aleatório seguro (secrets.randbelow()): {crypto_random}")

# Método 6: Usando o módulo uuid para gerar UUIDs
print("\nUsando o módulo uuid para gerar UUIDs")
uuid_value = uuid.uuid4()
print(f"UUID gerado (uuid.uuid4()): {uuid_value}")
```

Método $1$: Gera um número aleatório no intervalo $[0.0, 1.0)$ usando random.random()

Método $2$: Gera um número inteiro aleatório no intervalo $[1, 100]$ usando random.randint(a, b).

Método $3$: Gera um número aleatório com a semente (seed) definida como $42$ usando random.seed(seed).

Método $4$: Gera um número aleatório no intervalo $[0.0, 1.0)$ usando numpy.random.rand().

Método $5$: Gera um número aleatório seguro (criptograficamente seguro) no intervalo $[0, 100)$ usando secrets.randbelow(n).

Método $6$: Gera um UUID (identificador único universal) usando uuid.uuid4().

<br>

Os resultados desses métodos variam a cada execução do código, exceto quando uma semente (seed) é usada, o que pode tornar a geração de números pseudoaleatórios reproduzível. A utilização do módulo secrets garante que os números gerados são criptograficamente seguros, adequados para situações onde a segurança é crítica. O método uuid.uuid4() gera um identificador único universal que é extremamente improvável de se repetir, mesmo em diferentes sistemas distribuídos.


O código abaixo gera três conjuntos de dados, cada um seguindo uma distribuição diferente (normal, uniforme e Pareto), e plota histogramas para visualizar a distribuição de cada conjunto de dados.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Oct 23 07:23:37 2023

@author: amaury
"""
import numpy as np
import matplotlib.pyplot as plt

# Distribuição normal
mu, sigma = 0, 0.1  # Média e desvio padrão
normal_data = np.random.normal(mu, sigma, 1000)

# Distribuição uniforme
low, high = 0, 1  # Intervalo
uniform_data = np.random.uniform(low, high, 1000)

# Distribuição Pareto
a = 2.5  # Parâmetro de forma
pareto_data = np.random.pareto(a, 1000)

# Plot dos dados gerados
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='blue')
plt.title('Distribuição Normal')

plt.subplot(132)
plt.hist(uniform_data, bins=30, density=True, alpha=0.5, color='green')
plt.title('Distribuição Uniforme')

plt.subplot(133)
plt.hist(pareto_data, bins=30, density=True, alpha=0.5, color='red')
plt.title('Distribuição Pareto')

plt.tight_layout()
plt.show()
```

Os resultados gráficos são três histogramas, um para cada distribuição (normal, uniforme e Pareto), mostrando como os dados estão distribuídos em relação aos valores e suas frequências. O subgráfico "Distribuição Normal" exibe uma curva de sino característica. O subgráfico "Distribuição Uniforme" mostra uma distribuição uniforme retangular. O subgráfico "Distribuição Pareto" exibe uma cauda longa, indicando uma distribuição de Pareto. Esses gráficos ilustram visualmente as características de diferentes tipos de distribuições estatísticas.

O código a seguir realiza o ajuste de diferentes distribuições estatísticas (normal, exponencial e de Pareto) a um conjunto de dados aleatórios gerados.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Oct 23 07:34:12 2023

@author: amaury
"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, pareto
from scipy.optimize import curve_fit

# Gerar um conjunto de dados aleatórios
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# Criar um histograma dos dados
hist, bins = np.histogram(data, bins=30, density=True)
bin_centers = (bins[:-1] + bins[1:]) / 2

# Definir funções de ajuste para diferentes distribuições
def normal_distribution(x, mu, sigma):
    return norm.pdf(x, loc=mu, scale=sigma)

def exponential_distribution(x, scale):
    return expon.pdf(x, scale=scale)

def pareto_distribution(x, b):
    return pareto.pdf(x, b)

# Ajustar uma distribuição normal aos dados
params_normal, _ = curve_fit(normal_distribution, bin_centers, hist)
mu, sigma = params_normal
x = np.linspace(min(data), max(data), 100)
plt.plot(x, normal_distribution(x, mu, sigma), 'r-', label='Distribuição Normal')

# Ajustar uma distribuição exponencial aos dados
params_exponential, _ = curve_fit(exponential_distribution, bin_centers, hist)
scale = params_exponential
plt.plot(x, exponential_distribution(x, scale), 'g--', label='Distribuição Exponencial')

# Ajustar uma distribuição de Pareto aos dados
params_pareto, _ = curve_fit(pareto_distribution, bin_centers, hist)
b = params_pareto
plt.plot(x, pareto_distribution(x, b), 'm-.', label='Distribuição de Pareto')

# Exibir o gráfico com as distribuições ajustadas
plt.legend()
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.title('Ajuste de Diferentes Distribuições')
plt.show()

# Imprimir os parâmetros ajustados
print("\nParâmetros da Distribuição Normal (mu, sigma):", mu, sigma)
print("Parâmetros da Distribuição Exponencial (scale):", scale)
print("Parâmetros da Distribuição de Pareto (b):", b)
```

O gráfico exibe os dados originais e as curvas ajustadas para as distribuições normal, exponencial e de Pareto. Cada curva é plotada em uma cor diferente. A legenda indica qual distribuição corresponde a cada curva. Os parâmetros ajustados para cada distribuição também são impressos, mostrando os valores encontrados pelo processo de ajuste aos dados gerados. O ajuste de distribuições é uma técnica utilizada para modelar dados e entender a distribuição subjacente que melhor descreve o conjunto de dados observado.
