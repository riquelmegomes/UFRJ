# **Dynamic Mode Decomposition (DMD)**

> #### O método DMD (Dynamic Mode Decomposition) é uma técnica de análise de dados que é amplamente usada para extrair informações significativas de conjuntos de dados complexos, especialmente em aplicações relacionadas à dinâmica de sistemas. Ele foi desenvolvido na comunidade de dinâmica de fluidos, mas encontrou aplicações em uma variedade de campos, incluindo engenharia, física, biologia e finanças.

<br>

#### A ideia principal por trás do método DMD é decompor um conjunto de dados dinâmicos, geralmente representado por uma série temporal de observações, em modos dinâmicos (também chamados de modos DMD) que representam os padrões espaciais e temporais subjacentes nos dados. Esses modos são extraídos usando técnicas de álgebra linear e análise espectral.
<br>

#### Aqui estão os passos básicos do método DMD:
1. #### **Coleta de Dados:** Primeiro, você coleta um conjunto de dados que representa a evolução temporal de algum sistema dinâmico. Isso pode ser feito através de sensores, simulações computacionais, medições experimentais, etc.

2. #### **Divisão de Dados:** Os dados são divididos em duas séries temporais sobrepostas, geralmente chamadas de $X$ e $X'$. $X$ contém os dados em um dado momento, enquanto $X'$ contém os dados em um momento subsequente.

3. #### **Construção das Matrizes de Dados:** Com base nas séries temporais $X$ e $X'$, você constrói duas matrizes de dados, geralmente chamadas de $A$ e $B$, onde $A$ contém os dados em $X$ e $B$ contém os dados em $X'$. A matriz $B$ é uma versão deslocada no tempo da matriz $A$.

4. #### **Decomposição Espectral:** A próxima etapa envolve calcular a decomposição espectral da matriz $A$. Isso é feito usando a SVD (Decomposição em Valores Singulares), que fornece os modos DMD.

5. #### **Seleção de Modos DMD:** Os modos DMD representam as principais estruturas dinâmicas nos dados. Dependendo do problema, você pode escolher um número limitado de modos DMD para representar o sistema de maneira eficaz. Esses modos podem ser usados para prever a evolução futura do sistema, analisar suas propriedades de estabilidade e identificar padrões importantes.

6. #### **Reconstrução e Previsão:** Com os modos DMD selecionados, você pode reconstruir a evolução temporal do sistema ou fazer previsões futuras com base nesses modos.
<br>

#### O método DMD é especialmente útil quando se lida com sistemas dinâmicos complexos nos quais os modos subjacentes não são facilmente observáveis ou interpretáveis diretamente nos dados brutos. Ele oferece uma maneira eficaz de simplificar e compreender a dinâmica de tais sistemas, tornando-o uma ferramenta valiosa em uma variedade de aplicações científicas e de engenharia.

## **Princípios Básicos**

 - #### O método DMD é uma técnica de análise de dados que extrai modos dinâmicos a partir de observações experimentais ou simulações computacionais.
<br>

- #### Ele foi proposto em 2010 por Peter Schmid e Jonathan Tu, dois pesquisadores da área de fluidos.
<br>

- #### O DMD é aplicável em uma ampla variedade de áreas, incluindo engenharia, física, biologia, finanças, entre outras.
<br>

- #### O método DMD é baseado na decomposição em valores singulares (SVD) de uma matriz de dados $X$: $$X = UΣV^∗,$$ onde $U$ e $V$ são matrizes unitárias, e $Σ$ é uma matriz diagonal contendo os valores singulares.
<br>

- #### A matriz $X$ é formada por observações de uma variável ao longo do tempo.
<br>

- #### O objetivo do DMD é aproximar a matriz $X$ por uma matriz de baixo rank $X_r$, que contém os $r$ maiores valores singulares.

## **Operador de Koopman**

> #### O operador de Koopman é um conceito fundamental na teoria dos sistemas dinâmicos, utilizado para analisar e modelar a evolução temporal de sistemas complexos. Foi introduzido pelo matemático B.O. Koopman em 1931 e tem aplicações significativas em uma variedade de campos, incluindo física, biologia, engenharia e economia.

<br>

#### O operador de Koopman é uma representação matemática que descreve como as funções ou observáveis de um sistema dinâmico evoluem ao longo do tempo. Ele age sobre funções do espaço de estado do sistema e mapeia essas funções de um instante no tempo para outro.
<br>

#### O operador Koopman $U_t$ é definido como: $$U_tf (x_0) = f (T_t(x_0)),$$ onde $T_t$ é o operador de evolução temporal do sistema dinâmico e $f$ é uma função de observação.

<br>

#### **Observações**:
* #### O operador de Koopman é um operador linear que age em funções do espaço de estado de um sistema dinâmico e mapeia essas funções em outras funções do espaço de estado.

* #### Em outras palavras, o operador de Koopman descreve a evolução temporal de funções do espaço de estado sem a necessidade de resolver diretamente as equações diferenciais do sistema.

* #### O DMD pode ser visto como uma técnica para aproximar o operador de Koopman a partir de dados.

## **DMD Exata**

> #### Para um conjunto de dados dado por $X$, definimos o operador $A = YX^+$ onde $X^+$ é a pseudoinversa de $X$. A decomposição de modo dinâmico do par $(X, Y)$ é dada pela decomposição em autovalores de $A$. Ou seja, os modos e autovalores do DMD são os autovetores e autovalores de $A$.

<br>

#### **Observação**:
#### O operador A é a solução de mìnimos quadrados/mínima norma para o problema potencialmente sobre ou sub-determinado $AX = Y$. Ou seja, se existe uma solução exata para $AX = Y$ (o que sempre ocorre se os vetores $x_k$ forem linearmente independentes), então a escolha minimiza $||A||_F$ , onde $||A||_F = \sqrt{T_r(AA^∗)}$ denota a norma de Frobenius. Se não existe um $A$ que satisfaçaa exatamente $AX = Y$, então a escolha minimiza $||AX − Y ||_F$.

### dmdNovos4mejor.py


```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Mar 23 22:59:01 2023

@author: amaury
"""

import numpy as np
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# Criar dados sintéticos
t = np.linspace(0, 20, 200)
X = np.sin(2*np.pi*t/3) + np.sin(2*np.pi*t/6) + np.random.normal(0, 0.1, len(t))

# Formatar dados de entrada e saída
X1 = X[:-1]
X2 = X[1:]

# Transformar X1 para uma matriz (199, 2)
X1 = np.hstack((X1[:, np.newaxis], np.zeros((199, 1))))

# Aplicar a decomposição DMD
dmd = TruncatedSVD(n_components=2)
X1 = dmd.fit_transform(X1)

# reshape X2 to have two features
X2 = np.column_stack((X2, np.zeros(len(X2))))

# Extrair os modos
modes = dmd.transform(X2)

# Reconstruir os dados
X_dmd = dmd.inverse_transform(modes)

# Plotar os resultados
plt.plot(t, X, label='original')
plt.plot(t[1:], X_dmd, label='DMD')
plt.legend()
plt.show()
```

O gráfico gerado por este código mostra a série temporal original, representada pela linha azul "original", e a série reconstruída usando a Decomposição de Valores Singulares Truncados (Truncated SVD), representada pela linha laranja "DMD". O gráfico ilustra a capacidade da DMD em capturar as características principais da série temporal original, mesmo que a série seja uma combinação de duas funções seno e ruído.

Podemos observar que a série reconstruída segue de perto a série original, capturando a forma geral da série, mas não necessariamente os detalhes menores ou o ruído. Isso destaca a capacidade da Truncated SVD de reduzir a dimensionalidade dos dados, retendo as informações mais significativas.

O gráfico demonstra o potencial uso da DMD na análise e compressão de séries temporais, permitindo a representação de dados complexos em um espaço de menor dimensão, o que pode ser útil em várias aplicações, como análise de sinais, processamento de dados e redução de ruído.

### dmdmejor3.py

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Oct  2 00:16:02 2023

@author: amaury
"""

import numpy as np
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# Criar dados sintéticos
t = np.linspace(0, 20, 200)
X = np.sin(2 * np.pi * t / 3) + np.sin(2 * np.pi * t / 6) + np.random.normal(0, 0.1, len(t))

# Aplicar as funções seno e cosseno aos dados
X1 = np.column_stack((np.sin(X[:-1]), np.cos(X[:-1])))
X2 = np.column_stack((np.sin(X[1:]), np.cos(X[1:])))

# Aplicar a decomposição DMD separadamente para cada função
dmd1 = TruncatedSVD(n_components=2)
X1_dmd = dmd1.fit_transform(X1)

dmd2 = TruncatedSVD(n_components=2)
X2_dmd = dmd2.fit_transform(X2)

# Reconstruir os dados
X_dmd = np.column_stack((dmd1.inverse_transform(X1_dmd), dmd2.inverse_transform(X2_dmd)))

# Plotar os resultados
plt.plot(t, X, label='original')
plt.plot(t[1:], X_dmd[:, 0], label='DMD (sin)')
plt.plot(t[1:], X_dmd[:, 1], label='DMD (cos)')
plt.legend()
plt.show()
```

Aqui aplicamos as funções seno e cosseno aos dados. A primeira curva, rotulada como "original," representa a série temporal sintética original. Esta série é uma combinação de duas funções seno, com diferentes frequências, e um termo de ruído aleatório. A curva "original" é a base a partir da qual a análise é realizada. As duas curvas seguintes são as resultantes da aplicação da DMD separadamente para as componentes seno e cosseno da série temporal original. A curva "DMD (sin)" representa a componente seno da série, enquanto a curva "DMD (cos)" representa a componente cosseno.

O gráfico revela algumas informações importantes. Primeiramente, ele ilustra como a DMD pode separar eficazmente as componentes seno e cosseno da série original. A curva "DMD (sin)" captura as oscilações senoidais presentes nos dados, enquanto a curva "DMD (cos)" representa as oscilações cossenoidais. A sobreposição das curvas "DMD (sin)" e "DMD (cos)" em algumas partes do gráfico é esperada, uma vez que as funções seno e cosseno estão relacionadas por um deslocamento de fase de 90 graus.

Para modificar o código e aplicar outras funções aos dados, podemos escolher funções diferentes das senoides (seno e cosseno) e observar como a Decomposição de Valores Singulares Truncados (DMD) reage a essas funções. Aqui está uma versão modificada do código que aplica a função exponencial aos dados.

```python
import numpy as np
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# Criar dados sintéticos
t = np.linspace(0, 20, 200)
X = np.sin(2 * np.pi * t / 3) + np.sin(2 * np.pi * t / 6) + np.random.normal(0, 0.1, len(t))

# Aplicar a função exponencial aos dados
X_exp = np.exp(X)

# Criar uma matriz de características artificiais
X_artificial = np.column_stack((X_exp, X_exp))  # Aqui usamos duas colunas idênticas

# Aplicar a decomposição DMD
dmd = TruncatedSVD(n_components=2)
X_dmd = dmd.fit_transform(X_artificial)

# Reconstruir os dados
X_dmd_reconstructed = dmd.inverse_transform(X_dmd)

# Plotar os resultados
plt.plot(t, X_exp, label='original (exp)')
plt.plot(t[:len(X_dmd_reconstructed)], X_dmd_reconstructed[:, 0], label='DMD (exp)')
plt.legend()
plt.show()
```

Observa-se que a função exponencial amplifica a amplitude dos dados originais, fazendo com que a série se torne mais acentuada em comparação com a série original. A representação DMD (exp) captura os padrões e oscilações na série temporal exponencial. No entanto, como os dados de entrada para a DMD foram gerados a partir de uma matriz de características artificiais com duas colunas idênticas, as curvas da DMD e "original (exp)" acabam sendo bastante semelhantes.

O gráfico destaca a importância de escolher cuidadosamente a função de transformação aplicada aos dados antes de usar a DMD. A escolha da função de transformação pode afetar significativamente a interpretação e a eficácia da DMD na extração de padrões e características da série temporal original.

Aqui está um exemplo de código usando a Transformada de Fourier para análise da série temporal:

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar dados sintéticos
t = np.linspace(0, 20, 200)
X = np.sin(2 * np.pi * t / 3) + np.sin(2 * np.pi * t / 6) + np.random.normal(0, 0.1, len(t))

# Calcular a Transformada de Fourier dos dados
X_fourier = np.fft.fft(X)

# Frequências correspondentes
frequencies = np.fft.fftfreq(len(X))

# Filtrar componentes de alta frequência
X_filtered = np.copy(X_fourier)
X_filtered[np.abs(frequencies) > 0.1] = 0

# Reconstruir os dados com a Transformada Inversa de Fourier
X_reconstructed = np.fft.ifft(X_filtered).real

# Plotar os resultados
plt.plot(t, X, label='original')
plt.plot(t, X_reconstructed, label='Reconstruído (Frequências Filtradas)')
plt.legend()
plt.show()
```

Os resultados gráficos do código apresentam a série temporal original e a série reconstruída após a aplicação da Transformada de Fourier e a filtragem das componentes de alta frequência. Observa-se que a série temporal original possui várias oscilações de diferentes frequências, resultando em um sinal complexo.

A série reconstruída mostra como a filtragem das componentes de alta frequência afetou os dados originais. As componentes de alta frequência foram eliminadas, resultando em uma série suavizada que retém as principais oscilações de baixa e média frequência.

Por fim, vamos aplicar a função tangente hiperbólica (np.tanh) aos dados. Aqui está o código modificado:

```python
import numpy as np
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt

# Criar dados sintéticos
t = np.linspace(0, 20, 200)
X = np.sin(2 * np.pi * t / 3) + np.sin(2 * np.pi * t / 6) + np.random.normal(0, 0.1, len(t))

# Aplicar a função tangente hiperbólica aos dados
X1 = np.column_stack((np.tanh(X[:-1]), np.tanh(X[:-1])))
X2 = np.column_stack((np.tanh(X[1:]), np.tanh(X[1:])))

# Aplicar a decomposição DMD separadamente para cada função
dmd1 = TruncatedSVD(n_components=2)
X1_dmd = dmd1.fit_transform(X1)

dmd2 = TruncatedSVD(n_components=2)
X2_dmd = dmd2.fit_transform(X2)

# Reconstruir os dados
X_dmd = np.column_stack((dmd1.inverse_transform(X1_dmd), dmd2.inverse_transform(X2_dmd)))

# Plotar os resultados
plt.plot(t, X, label='original')
plt.plot(t[1:], X_dmd[:, 0], label='DMD (tanh)')
plt.plot(t[1:], X_dmd[:, 1], label='DMD (tanh)')
plt.legend()
plt.show()
```

A suavização na série reconstruída pode ser atribuída à não linearidade introduzida pela função tangente hiperbólica. A reconstrução destaca as características globais da série original, enquanto detalhes menores podem ser perdidos.

Em resumo, o gráfico demonstra a aplicação da função tangente hiperbólica aos dados e como a Truncated SVD é usada para reconstruir a série temporal após a transformação. A escolha da função e do método de decomposição depende dos objetivos específicos da análise de dados.
