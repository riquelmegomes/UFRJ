```julia
# Republicanos e monarquistas

function calcula(k)
    r0, m0 = 50, 50
    r, m = 0, 0
    
    # println("Ano 0: ", (r0, m0))
    for i=1:k
        r = 0.7*r0 + 0.2*m0
        m = 0.3*r0 + 0.8*m0
        r0, m0 = r, m
        # println("Ano ", i, ": ", (r, m))
    end
    
    return r, m
end
```

```julia
calcula(10)
```

```julia
calcula(52)
```

```julia
# Bactérias

function sistema_pop(t, p0)
    p = zeros(4)
    T = [0 0.5 0.2; 0.8 0 0; 0 0.5 0]
    
    println("Mês 0: ", p0)
    for i=1:t-1
        p = T*p0
        p0 = p
        println("Mês ", i, ": ", p)
    end
    
    p = T*p0
    println("Mês ", t, ": ", p)
    return p
end
```

```julia
p0 = [250; 0; 0]
sistema_pop(5, p0)
```

```julia
sistema_pop(1000, p0)
```

```julia
p0 = [50; 25; 25; 0]
sistema_pop(1000, p0)
```

```julia
function pagerank(t, p0)
    p = zeros(4)
    T = [0 0 1 1/2; 1/3 0 0 0; 1/3 1/2 0 1/2; 1/3 1/2 0 0]
    
    println("Clique 0: ", p0)
    for i=1:t
        p = T*p0
        p0 = p
        # println("Mês ", i, ": ", p)
    end
    
    println("Clique ", t, ": ", p)
    return p
end
```

```julia
p0 = [25; 25; 25; 25]
pagerank(100, p0)
```

```julia
p0 = [0; 0; 0; 100]
pagerank(100, p0)
```

```julia
function vampiros(t, p0)
    p = zeros(2)
    T = [1.01 0; 0.01 0.99]
    
    println("Ano 0: ", p0)
    for i=1:t
        p = T*p0
        p0 = p
        #println("Ano ", i, ": ", p)
    end
    
    println("Ano ", t, ": ", p)
    return p
end
```

```julia
p0 = [100, 0]
vampiros(100, p0)
```
